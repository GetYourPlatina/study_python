```python
# 객체(Object)에 대한 이해
# a=[1,2,3] 이라는 list 변수를 생성해보자.
# a 변수에 대응하는 [1,2,3]은 메모리에 저장되는데 이 [1,2,3] 자체가 객체이다. (자료형이 리스트인 객체이다.)
# 객체에는 속성(attribute, 또는 값(value))과 메서드(method)가 포함되어 있다.
# 속성은 해당 객체가 가지는 고유의 특성이나 상태 등을 의미한다.
# 메서드는 해당 객체에 할당된 정의된 행동(또는 할당된 함수)를 의미한다.

# 자 객체를 만들어보자!
a=[1,2,3]

# dir 함수를 쓰면 해당 객체에 포함된 속성 그리고 메서드가 전부 표시된다.
dir(a)

# 재미있는 것은 메서드 종류이다. add, mul 등의 연산은 물론 앞서 활용한 dir 함수도 포함되어 있다.
# 그렇다! 객체지향 프로그래밍이란 이런것이다!
# 객체에 속성은 물론 연산 등의 메서드가 모두 객체에 할당되어있고, 이를 불러와서 동작하는 것이다.
# 예를 들어보겠다. a=2,b=7로 정의하고 a+b를 진행해보자
a=2;b=7
c=a+b; print(c) # 그러면 9가 나온다.. 당연하지..? 근데 이 연산 객체지향 프로그래밍인 python에선 사실 아래처럼 동작한다!
d=a.__add__(b); print(d) #어떤가 값이 똑같지 않은가?

# 이처럼 객체에는 정의된 행동인 method가 할당되어 있고 이를 활용하여 동작이 이뤄지는 것을 알 수 있다.
# 즉 데이터도, 행동도 모두 객체에 비롯된다(객체로부터 시작된다).

# 턴더 메소드(dunder method(double-under method)) 또는 special method
# 턴더 메소드는 일반 method와는 다르게 python 인터프리터가 자동으로 호출하는 특별한 메서드 입니다.
# 특정 python 문법을 사용하면 이에 대응하는 special method가 자동으로 호출되는 것입니다.
# ex) a+b를 쓰면 a.__add__(b)가 자동으로 호출되는 것
# ex) for x in a를 쓰면 a.__iter__(~~)를 호출하는 것
#
# 질문! 그러면 더블언더스코어(__)가 들어간 것은 모두 턴더 메소드인가요?
# 대답은 No!! special attribute도 존재합니다! __init__ 이나 __class__ 등이 그렇죠..
# 주의할 점!! 더블언더스코어는 앞으로 배울 class의 attribute나 method를 정의할 때 쓰지 말아달라고 부탁합니다! 왜냐하면 python이 버전업이 되면서 추가되는 special method를 정의하는데 활용되어야 하기 때문이죠.
# 어떻게 보면 턴더 메소드는 python에서 기본적으로 포함되는 메소드라고도 할 수 있겠네요..
```

```python
# class란 무엇인가?
# 자 그러면 class는 무엇이에요? 바로 속성과 메서드를 하나의 네임스페이스로 묶는 객체!이며, 새로운 인스턴스를 생성하기 위한 호출 가능한 객체이다!
# 즉 동일한 객체를 생성하기 위해 설계(정의가 맞는 표현이지만 직관적으로는 이 표현이 이해가 쉬울듯...)된 하나의 틀이라고 보는 것이 맞을듯 하다.
# 재미있는 사실은 위에서 밝혔듯이 객체를 생성하는 객체이다!!

# calculator class 객체를 생성해보자 
class Calculator:
    def __init__(self):
        self.result = 0

    def add(self, num):
        self.result += num
        return self.result

# 위에 생성된 class를 이용하여 2개의 인스턴트(객체)를 생성해보면 다음과 같다.
cal1 = Calculator()
cal2 = Calculator()

print(cal1.add(3))
print(cal1.add(4))
print(cal2.add(77))
print(cal2.add(99))

# 재미있는 사실은 class로 생성된 인스턴스(객체) cal1과 cal2는 각각의 객체로 상호간에 간섭을 받지 않는다. 다만 class에 정의한 것과 동일한 구조를 가지고 있다.

# 자 여기서 내가 class로 생성된 객체를 인스턴스라 표현하였는데, 정확하게 짚고 가자.
#  - class로 만든 객체를 인스턴스라 한다. 이를 표현할 때 예시는 아래와 같다.
#  - Calculator()로 만든 cal1 또는 cal2는 객체이다. 
#  - cal1 또는 cal2는 Calculator의 인스턴스이다.
# 즉 인스턴스라는 말은 특정 객체(a)가 어떤 class의 객체인지 관계 위주 설명 시에 사용한다.


# 그렇다면 궁금한 점!
d1=[1,2,3] #으로 생성되는 리스트 이것도 class인 것이고 이를 호출하는 걸까?
# 답은 반은 맞고 반은 틀리다. list는 클레스가 맞다!! 근데 호출방식이 좀 다르다!
# 왜냐면 []는 list로 정의된 literal(변하지 않는 데이터, 프로그래밍 언어의 소스 코드에서 특정한 자료형의 값을 그대로(literally) 표현하는 문법 또는 토큰을 말한다.)
```

```python
# Class 실습으로 간다! (class 구조 만들기!)

# class 구조 만들기
# 아래는 아무것도 수행하지 않는 문법으로, 임시 코드를 작성할 떄 주로 사용한다.
class FourCal:
    pass

# 위의 class도 class 객체인 만큼 객체를 생성할 수 있다. 위의 FourCal class로 생성한 인스턴스 a의 type를 확인하면 다음과 같다.
a=FourCal()
print(type(a))
# type를 진행하면 어느 class의 인스턴스 인지 확인도 가능하다.
```

```python
# Class 실습으로 간다! (class 객체에 연산할 숫자 지정하기)
# 이제 연산을 수행할 수 있는 class를 만들어 보겠다.
# 그런데 사칙연산을 수행할때 사칙연산을 진행할 대상(매개변수)을 객체에 지정할 수 있도록 아래와 같이 만들어보자.
# 위의 FourCal class에 pass를 삭제하고 아래와 같이 data를 setting 하는 함수를 넣어보자

class FourCal:
    def setdata(self,first,second):
        self.first=first
        self.second=second

# 여기서 setdata 함수의 첫번쨰 매개변수가 self이다. 왜지? 
# 일단 위의 class를 호출해서 객체를 만들고 setdata 매서드를 호출해보자
a=FourCal()
a.setdata(11,99)

# 위와 같이 매개변수(11,99)를 입력해서 객체의 매서드를 호출하면 입력된 매개변수는 정의된 객체 내부 매개변수에 대응하여 입력된다.
# a.setdata(11,99) ---->> def setdata(self,first,second)
# a -> self 
# 11 -> first
# 99 -> second
#
# 즉 setdata의 첫번째 입력매개변수인 self에는 setdata 메서드를 호출한 객체 a가 자동으로 전달된다.
# 이는 파이썬의 독특한 특징으로 자바 같은 언어에서는 이 self를 쓰진 않는다..

# 메서드를 호출하는 방법
# 잘 사용하지는 않지만, 다음과 같이 클래스를 이용해 메서드를 호출할 수도 있다.
a=FourCal()
FourCal.setdata(a,11,99)
# 이처럼 class 명을 다이랙트로 호출해서 매서드를 호출할 수도 있지만 이런 경우에는 생성한 변수(class로 생성한 인스턴스)를 꼭 지정해줘야 한다.
a=FourCal()
a.setdata(11,99) # 생성된 객체의 변수로 호출하면 self는 넣을 필요가 없다...


# 매개변수 입력 방법 알아보기
# 매개변수를 입력하는 방법은 다양하다.
# 위와 같이 변수.매서드를 호출하여 매개변수를 전달하는 방법이 있다.
a.setdata(11,99)
print(a.first)
print(a.second)
# 매서드안에 정의된 변수를 직접 입력하는 방법도 있다.
a.first=1
a.second=5
print(a.first)
print(a.second)

# 잠깐? setdata를 하기 전에도 위의 방법은 통하는가?
b=FourCal()
b.first=1
print('after change: '+str(b.first)) #통한다! 즉 setdata의 매서드를 쓰지 않아도 객체에 first 값이 정의되었다.
# 위의 b.first=1를 수행하면 b의 객체에 객체변수인 first가 생성된다! 
# 이렇게 객체내 생성되는 객체변수를 인스턴스 변수 또는 속성(attribute)라고 한다.

# dir을 쓰면 객체 내 method와 attribute를 볼수 있는 것을 알수 있을것이다! 확인해보자!
# first가 있는가!?
dir(b)
# 우왕! 있다! 신기방기!

# 자 그러면 first의 값을 확인해보자!
print(b.first)

# 단! class로 변수를 생성한 직후 first 값을 확인하고자 하면 그것은 에러가 뜬다. 
# 왜냐하면 first 값이 아직 객체에 정의되지 않았기 때문이다. 즉 해당 객체에 속성인 first가 정의되지 않음!
```

```python
# Class 실습으로 간다! (operator)

class FourCal:
    def setparm(self,x,y):
        self.first=x
        self.second=y
    def add(self):
        result=self.first+self.second
        return result
    def mul(self):
        result=self.first*self.second
        return result
    def sub(self):
        result=self.first-self.second
        return result
    def div(self):
        if self.second == 0:
            print('cal error is occurred. plz check denominator number. maybe denominator is zero.',end=' results: ')
        else:
            result=self.first/self.second
            return result
        
# 만든 class를 활용해보기!
a=FourCal()
a.setparm(11,99)
result1=a.add()
print(result1)
result1=a.mul()
print(result1)
result1=a.sub()
print(result1)
result1=a.div()
print(result1)

print('')

a.setparm(11,0)
result1=a.add()
print(result1)
result1=a.mul()
print(result1)
result1=a.sub()
print(result1)
result1=a.div()
print(result1)
```

```python
# 생성자에 대해서 알아보자
# 위의 FourCal class를 호출하고 method인 add를 호출하면 애러가 발생한다!
# 그 이유는 당연 속성인 first와 second가 정의되지 않았기 때문이다.
# 이 때문에 생성자(constructor)를 구현하여 class 객체를 생성할 때 first와 second를 정의할 수 있도록 하는 것이 좋다.
 
# 생성자를 설정하는 방법은 다음과 같은 매서드를 추가하여 진행할 수 있다.
# __init__의 매서드를 추가하면 class 생성 시 자동으로 호출된다.
# 즉 생성자란 객체가 생성될 때 자동으로 호출되는 메서드를 의미한다.

class FourCal:
    def __init__(self,first,second):
        self.first=first
        self.second=second
    def add(self):
        result=self.first+self.second
        return result
# 위의 클래스를 생성하고 a 객체를 다음과 같이 생성해보자
# a=FourCal() 생성이 되지 않는다! 다음과 같이 에러가 뜬다!
# TypeError: FourCal.__init__() missing 2 required positional arguments: 'first' and 'second'
# 그렇다! 생성자를 생성했으니, 이에 맞는 매개변수(속성이 될 값 first와 second를 입력해줘야 한다.)
# 따라서, 정상적으로 생성하려면 다음과 같다.

a=FourCal(55,66)
print(a.first)
print(a.second)
# 잘 생성된 것과 first 그리고 second의 속성도 잘 생성된 것을 확인할 수 있다.
# 위의 setdata와 완전 동일하나, __init__으로 매서드를 정의해서 생성자로 자동 인식되고, 객체 생성시 자동 호출되는 것이다.

# 자 그러면 생성자로 정의된 속성을 활용하여 다른 매서드 인 add를 구해보자
results=a.add()
print(results)
```

```python
# 클래스의 상속(Inheritance)
# 어떤 클래스를 만들때 다른 클래스의 기능을 물려받을 수 있게 만드는 것을 상속이라 한다.
# 상속하는 방법은 다음과 같다.
class test_FourCal(FourCal):
    pass
# 즉 "class 클래스_이름(상속할_클래스_이름)" 을 하면 상속이 가능하다.
# 상속을 받은 test_FourCal class는 상속된 class인 FourCal의 속성과 매서드 역시 그대로 상속 받는다.

a=test_FourCal(55,21)
results=a.add()
print(results)
print('')
# 상속받은 class의 기능이 잘 사용되는 것을 확인할 수 있다.

# 왜 상속을 쓸까?
# 1. 기존의 class를 유지하되 추가적인 매서드나 속성을 부여하고 싶으니까!
# 2. 기존의 class를 유지하되 class 내 매서드나 속성을 수정하고 싶으니까!

# 자 그럼 add를 한번 바꿔보자!
class test_FourCal(FourCal):
    def add(self):
        result=self.first+1
        return result

a=test_FourCal(55,21)
results=a.add()
print(results)
# 이렇게 진행하니 add 매서드의 기능이 수정되었다!
# 이처럼 add 매서드를 다시 쓰는 경우를 메서드 오버라이딩(method overriding) 이라고 한다!
# 오버라이딩 된 매서드가 있다면 부모 클래스의 메서드 대신 오버라이딩한 메서드가 호출된다!

# 이처럼 메서드의 상속은 메서드 기능의 확장할 때 주로 사용한다.
```

```python
# Class 변수
# 클래스 변수는 클래스 안에서 변수를 선언하여 생성한다. 
class family:
    last_name='kim'

a=family()
print(a.last_name)
# class 변수는 선언한 class에 모두 공유되는 변수이다. 따라서 class에서 생성한 인스턴스는 생성된 클래스 변수를 모두 가지게 된다.
# 그런데! 동일한 변수명으로 객체변수(속성)을 생성하면 덮어씌워진다.
a.last_name='good'
print(a.last_name)
print('')

# 새로 생성한 b 변수의 class 변수도 비교해보자!
b=family()
print(b.last_name)
print('')
# 위에서 선언한 클래스 변수의 객체가 반환된다.

# 마지막으로 클래스 변수를 변경하면 어떻게 될까?
c=family()
family.last_name='park'
print(a.last_name)
print(b.last_name)
print(c.last_name)

# a는 객체변수(속성)을 선언했으므로 객체변수의 값이 호출되었다.
# b와 c는 클래스 변수가 호출되었으며 일괄적으로 변경이 진행되었다!
# 이처럼 클래스 변수가 변경되면, 생성된 class 인스턴스가 가진 클래스 변수 역시 모두 변환된다!
```
