```python
# 1st Number(숫자형)
# 1. 정수형(Integer) - 정수를 뜻하는 자료형(양의정수,음의정수,숫자 0)
a = 79
b = -77
c = 0

# 2. 실수형(floating-point) - 소수점이 포함된 숫자형(-infinity ~ infinity)
a = 3.1415926
b = -1.41421356

# 3. 8진수(Octal) - 8진수으로 표현되는 숫자형, 0o 또는 0O로 시작한다. (형태: 10진수에서 82는 8진수에서 0o122 or 0O122)
a = 0o122
b = 0O122
print("1. Octal example")
print("small 0o122 "+str(a))
print("is same")
print("capital 0O122 "+str(b)+"\n")

# 4. 16진수(hexadecimal) - 16진수로 표현되는 숫자형, 10부터는 a,b,c,d,e,f로 표기, 0x로 시작한다. (형태: 0x8ff)
a = 0x8ff
print("2. Hexadecimal example")
print("Decimal 2303 is same hexadecimal 8ff")
print(str(a) + "\n")

# 5. 사칙연산(Elementary arithmetic)
print("3. Elementary arithmetic example")

a = 3
b = 4
c = a + b
print("3.1. plus(+): a=3, b=4, a+b=? "+str(c))

a = 3
b = 4
c = a - b
print("3.2. minus(-): a=3, b=4, a-b=? "+str(c)+"\n")

# +예제! Integer과 floating-point를 연산하면?
a = 3.1415926
b = 4 
c = a + b
print("bouns 1. Is it possible to add integer and floating-point? \nplus: a=3.1415926, b=4, a+b=? "+str(c)+"  result: possible!")
print("what is the format of result?  " + str(type(c))+"\n")
# 정수끼리의 연산으로 정수가 결과가 나오면 정수가 되지만 정수가 아닌 실수가 나오면 실수로 자동 변환됨!

a = 3
b = 4
c = a * b
print("3.3. multiplication(*): a=3, b=4, a*b=? "+str(c))

a = 3
b = 4
c = a / b
print("3.4. division(/): a=3, b=4, a/b=? "+str(c))

a = 3
b = 4
c = a**b
print("3.5. square(**): a=3, b=4, a^b=? "+str(c))

a = 13
b = 4
c = a%b
print("3.6. remainder(%): a=13, b=4, a%b=? "+str(c))
#remainder는 나머지! 나눗셈을 했을때 나머지로 남는 값을 의미

a = 13
b = 4
c = a//b
print("3.7. quotient(//): a=13, b=4, a//b=? "+str(c)+"\n")
#quotient는 몫! 나눗셈을 했을때 몫을 의미!

# complex operator
a = 1
a = a + 1 # 기존 변수에서 1이 증가하거나
a = a - 1 # 기존 변수에서 1이 마이너스 되는 경우
# 아래와 같이 간략하게 표현할 수 있음
a = 1
a+=1 #이것은 위의 a = a + 1과 같음
a-=1 #이것은 위의 a = a - 1과 같음

# 비교
print("bouns 2. Compare normal plus and complex operator")
a = 1
a = a + 1
print(str(a)+" was performed by a=a+1.")
a = 1
a+=1
print(str(a)+" was performed by a+=1. Same!!")

# Complex operator 종류는 다양하다!
# +=, -=, *=, /=, //=, %=, **=
# 예제는 위와 같으니 생략!

# 숫자형 Done
```

```python
# 2nd String(문자열)
# 문자열을 쓰는법!!
"Hello world!"
'Hello world!'
"""Helllllo world"""
'''Hellow world!'''
test='hello world'
print('1. welcome to String world: ' + test)

# 왜 마지막 글자만 나올까? 그것은 print로 쓰지 않았기 때문!
# 여튼 큰따옴표, 작은따옴표, 큰따옴표 3개, 작은따옴표 3개 모두 사용 가능하다!


# 만약 작은따옴표를 글에 넣고 싶으면 큰따옴표를 쓰면 되고, 반대로 큰따옴표를 넣고싶으면 작은따옴표를 쓰면된다.
print('2. how to use quotation marks?')
test="Hello 'kim'"
print('  Case inside single quotation marks: '+test)
test='Hello "kim"'
print('  Case inside Double quotation marks: '+test)

# 다른방법! 역슬래시(\)를 이용하면 된다! \' 또는 \"
print('2.2 how to use quotation marks? - 2')
test="Hello \'k\"i\'m"
print('  Case inside single quotation marks: '+test)

# 줄바꾸기 방법: 첫번쨰는 이스케이프 코드를 쓰면 된다.
print('3. newline method')
test='Hello \n  world'
# 다른 하나는 따옴표(크든 작든) 3개를 연속으로 쓰면 줄을 띄어 써도 적용된다!!
print('  Newline!(escape code ver): '+test)
test="""Hello
  world"""
print('  Newline!(quotation marks ver): '+test)

# 문자열 연산하기!
# 문자열도 더하거나 곱할수 있다는거!(파이썬만 가능하오!)
print('4. str plus and multiply')
head='hello '
tail='world'
print('  str plus(hello + world): '+head+tail)
print('  str multiply(hello * 2): '+head*2)

# power 응용!
print('   example) many programmer use this technique')
print('   '+'='*50)
print('   Hello world!!')
print('   '+'='*50)

# 문자열 인덱싱과 슬라이딩
# 인덱싱 하고자 하는 문자의 위치를 찾아 해당 문자 변수에 대괄호([])로 위치를 지정하면 됨
print('5. str indexing and slinding')
data='Hello my world. I\'m your GOD'
print('  str index(m): '+data[6]) #순방향 인덱스
print('  str index(O): '+data[-2]) #역방향 인덱스
print('  str index using sliding technique(my): '+data[6:8])
print('  str index using sliding technique(Hello): '+data[:5])
print('  str index using sliding technique(GOD): '+data[-3:])
print('  str index using sliding technique2(GOD): '+data[len(data)-3:len(data)])

# 중요한 부분! 문자열은 변경 불가능한 변수이므로 아래와 같이 변경을 해줘야 한다.
print('6. revised string')
data='pithon'
renewal_data=data[0]+'y'+data[2:]
print('  old_word: '+data+', revised_word: '+renewal_data)

# String formatting
# 문자열 안에 다른 문자열 또는 숫자열 등의 다른 값을 삽입하는 방법!
# 숫자열은 문자 안에 %d, 문자열은 문자 안에 %s를 넣어주면 된다.
# 복수의 경우 %뒤에 (변수1,변수2,...)을 넣어주면 된다.
print('7. String formatting')
tem=20
weather='sunny'
disp='  Today Temperture is %d.' % tem
print('  single string formatting: ' + disp)
disp2='  Today Temperture is %d and %s' %(tem,weather)
print('  multi formatting: ' + disp2)
# 변수는 물론 다이렉트로 변수 대입도 가능하다.
print('  direct formatting(number): i have %d dollers' % 50e100)
print('  direct formatting(string): i have %s dollers' % '500billion')
# format code(포맷코드)는 다음과 같다.
# %s: string, %c: characterm, %d: integer, %f: floating-point, %o: octal, %X: hexadecimal, %%: %(%를 표시하기 위해선 %를 두번해주자! 약속!)
# %s는 숫자를 integer를 넣던 floating-point를 넣던 문자열로 변환하여 입력된다!
# 그리고 %d와 %f는 구분을 잘해줄것!

# formatting 관련 추가 공부내용
# 1) 정렬과 공백
# formatting에서 사용하는 포맷코드의 사이에 숫자를 함께쓰면 입력 숫자 자리수를 지정할 수 있고 이를 이용하여 공백을 쓸수 있다.(지정입력수 대비 실제입력수가 적으면 공백으로 남음)
# 예를들어 %10s를 쓰면 10개의 자리에서 숫자는 오른쪽으로 정렬하고 숫자를 입력하되 남는 공간은 공백으로 두라는 뜻이 된다.
print('8. formatting blank and sorting')
print('  Blank 10 block1: %10sgood' % 500) # 3개 자리 지정 후 7개 숫자만 들어가서 빈 5자리는 공백이 된다.
print('  Blank 10 block2: %10sgood' % 55500) # 10개 자리 지정 후 5개 숫자만 들어가서 빈 5자리는 공백이 된다.
print('  Blank 10 block3: %10sgood' % 1000055500) # 10개 자리 지정 후 10개 숫자가 들어가면 꽉차게 들어가서 공백이 없다.
print('  Blank 10 block3: %10sgood' % 10000055500) # 공백이 초과되면 그냥 일반적인 입력과 다름 없다.
# 정렬을 왼쪽으로 하고 싶으면 지정숫자 자리수에 -를 추가하면된다. ex) %-10s를 하면 입력 숫자는 왼쪽에 정렬됨
print('  Blank 10 block1: %-10sgood' % 500) # 3개 자리 지정 후 7개 숫자만 들어가서 빈 5자리는 공백이 된다.
print('  Blank 10 block3: %-10sgood' % 1000055500) # 10개 자리 지정 후 10개 숫자가 들어가면 꽉차게 들어가서 공백이 없다. 그래서 정렬이 쓸모가 없다.
# 2) 소수점(decimal point) 표현하기
# formatting을 이용하여 floating-point 형식의 숫자를 입력하면 다음과 같음(포맷코드: %f)
print('9. decimal point')
# 앞서 정렬과 공백과 유사하다. 포맷코드 사이 숫자를 소수점으로 넣어주면 된다.(참고로 반올림도 한다)
print('  decimal point_hundredths place value(0.01),: %0.2fgood' % 3.1415926) # 3개 자리 지정 후 7개 숫자만 들어가서 빈 5자리는 공백이 된다.
print('  decimal point_ten-thousandths place value(0.0001),: %0.4fgood' % 3.1415926) # 3개 자리 지정 후 7개 숫자만 들어가서 빈 5자리는 공백이 된다.
# 여기서 소수점 아래(오른쪽에 위치한 숫자)의 숫자는 자릿수를 의미하며 2면 2자리까지 10이면 10자리까지 표기하도록 하는것이다.
# 이때 소수점 왼쪽 정수자리는 여전히 자릿수 입력으로 이를 같이 활용할 수도 있다. 예를 들면 10.4라 입력하면 10자리수 입력하고 소수점 4자리까지를 의미함
print('  Blank10 + decimal point(2),: %10.2fgood' % 3.1415926) # 3개 자리 지정 후 7개 숫자만 들어가서 빈 5자리는 공백이 된다.
print('  Blank07 + decimal point(2),: %7.2fgood' % 3.1415926) # 3개 자리 지정 후 7개 숫자만 들어가서 빈 5자리는 공백이 된다.
print('  Blank03 + decimal point(2),: %7.2fgood' % 123.1415926) # 3개 자리 지정 후 7개 숫자만 들어가서 빈 5자리는 공백이 된다.
print('  Blank03 + decimal point(4),: %7.4fgood' % 123.1415926) # 3개 자리 지정 후 7개 숫자만 들어가서 빈 5자리는 공백이 된다.

# format 함수를 이용한 formatting
print('10. formatting using format function')
# format 함수를 이용한 formatting도 가능하며, 숫자로 순서를 이용한 포맷팅이나 문자열로 지정하는 포맷팅 모두 가능하다. format 함수 포맷팅은 중괄호를 쓴다.
# format을 쓸땐 .format으로 쓸것. (문자열 쓸때 내부 함수인듯?)
print('  format function using Number: you have a {0}'.format(1))
print('  format function using Number: you have a {1} and {0}'.format(1,10)) # 중괄호 안에 숫자는 포맷팅 순서를 의미하며 0부터 시작한다.
print('  format function using String: you have a {1} and {0}'.format('bottom','cup'))
print('  format function using String and Number: you have a {1} and {0}'.format(0,'cup')) # 숫자와 문자를 복합적으로도 포맷팅할 수 있다.
number=321
print('  format function using variable: you have a {1} and {0}'.format(0,number)) # 변수를 포맷팅 할수도 있다!
# 포맷팅에 숫자 순서가 아닌 변수명을 직접 지정할 수도 있는데 이때는 format 함수 내에 값도 있어야 한다. 외부 변수를 끌고 오는 경우는 위의 사례처럼 사용해야 함
print('  format function using variable direct and String: you have a {num} and {0} and {number}'.format(0,number=321,num=999)) 
# 단 다음의 경우는 동작 안됨! 
# 1. print('you have a {1} and {0}'.format(0,number=0)) format 함수에 변수=값으로 입력되면 변수 이름으로 포맷팅해야됨. 단순 위치로를 실패!
# 2. number=1로 변수가 있는경우 print('you have a {number} and {0}'.format(0,number))

# formatting 함수 내에서도 자릿수를 설정할 수 있고, 정렬을 할수 있다. 방식은 아래와 같다.
print('  format function blank 10 and center sort: you have a {0:^10}'.format('hi'))
print('  format function blank 10 and left sort: you have a {0:<10}'.format('hi'))
print('  format function blank 10 and right sort: you have a {0:>10}'.format('hi'))
# 또한 공백을 채울수도 있다!
print('  format function blank 10 and left sort: you have a {0:=<10}'.format('hi'))
print('  format function blank 10 and center sort: you have a {0:=^10}'.format('hi'))
# 소수점은 다음과 같이 표현할 수 있다. (위랑 유사하다! 다만 0.2 의 자릿수 설정 옆에 f를 추가해줘야 한다. ex) 10.2f - 자릿수 10에 소수점 자리 2)
print('  format function blank 10 and left sort: you have a {0:=<10.2f}'.format(3.1415))
print('  format function blank 10 and center sort: you have a {0:=^10.2f}'.format(3.1415))

# 엇? 그러면 중괄호를 문자로 쓰고 싶으면 어떻게요? 이들은 걍 반복으로 쓰면 문자로 인식하게 규칙을 정했다! %도 그렇지 않은가?
print('10. Curly brackets')
print('  brackets! yes format: {{goood}}'.format())
# 엇 근데 format 함수 안쓰면요? 엄... 그럼 걍 상관 없어용
print('  brackets!  no format: {{goood}}')
print('  brackets!  no format: {goood}')

print('\n') #make blank

# f 문자열 포매팅 %중요! python 3.6버전 이후부터 사용가능하며 이전 버전에서는 사용 불가능
print('11. fString formating')
name = 'HbKim'
age = 36
print(f'  fString formatting: my name is {name}. and my age is {age}')
print(f'  fString formatting with operator: my name is {name}. and my age is {age+1}')
# 뒤에 나올 딕셔너리 형태 변수를 포맷팅 할 수도 있다.
d={'name':'HbKim','age':30}
print(f'  fString formatting with dictionary value: my name is {d["name"]}. and my age is {d['age']} or {d['age']+1}') # 이 역시도 연산은 먹는다.
# 정렬과 소수점은 다음과 같고 유사하니 참고 바란다.
print(f'  fString blank and sorting: {'hi':^10}')
print(f'  fString blank and sorting: {'hi':=^10}')
print(f'  fString blank and sorting: {age:^10}')
pi=3.1415926
print(f'  fString blank and sorting: {pi:^10.5f}') #f를 까먹지 말자
# f 포맷팅은 화폐 단위의 콤마를 삽입하는 기능도 있으니 참고하라
money=5000000000000000000000000 # 5e100와 같이 표기는 콤마 안먹으니 참고!
print(f'  I have {money:,} dollers')


# 문자형 기본 설명! Done!
```

```python
# 문자열 자체가 가진 함수들 정리!
# 1. count - 문자 개수 세기
data = 'Faker is good'
print('ex word is \''+data+'\'')
print('1.1 total word count: '+str(data.count('')))
print('1.2 partial word count(f): '+str(data.count('f')))
print('1.3 partial word count(F): '+str(data.count('F'))) # 대소문자를 가려서 찾는다!
print('')

# 2. find - 문자의 글자 위치 알려주기
print('2.1 find(a): '+str(data.find('a')))
print('2.2 find(go): '+str(data.find('go'))) #복수의 경우 가장 첫글자 위치 알려줌
print('2.3 find(god): '+str(data.find('god'))) #없는 경우는 -1로 표기
# 3. index - 문자의 위치 찾기 2
print('3.2 index(a): '+str(data.index('a')))
print('3.3 index(go): '+str(data.index('go'))) #복수의 경우 가장 첫글자 위치 알려줌
# index는 없는 경우 -1이 아닌 바로 오류를 발생시킴. 이점이 다르다!
print('')

# 4. join - 문자열 삽입: 문자열을 삽입하는 데, 아래와 같다. 보면서 확인
print('4.1 join: '+','.join(data))
print('4.2 join: '+',!'.join(data))
print('4.3 join: '+data.join('!@!'))
# 보면 알다시피 join 안의 문자 사이사이에 join 앞 문자나 변수를 끼워 넣는것!
print('')

# 5. upper - 소문자를 대문자로 바꾸는 것, lower = 대문자를 소문자로 바꾸는 것
high_data = data.upper()
print('5.1 upper: '+high_data)
print('5.2 lower: '+high_data.lower())
print('')

# 6. lstrip, rstrip, strip - 공백지우기!
data = '  Faker is good  '
print('6. erase ex word is \''+data+'\'')
print('6.1 erase left: '+data.lstrip())
print('6.2 erase right: '+data.rstrip())
print('6.3 erase left and right: '+data.strip())
data = 'Faker is good'
print('')

# 7. 문자열 바꾸기 = replace(바꾸고 싶은 단어,바꿀단어)
print('7 replace: '+data.replace('Faker','ruler')) #없는단어 치면 오류 나옴 ㅠ
print('')

# 8. 문자열 나누기 = split: 띄어쓴 부분을 모두 나눠버렷!
split_data=data.split()
print('8. split: '); print(split_data)
print('')

# 9.1 알파벳 검사 = isalpha(알파벳으로만 구성되어 있나요?)
print('9.1 isalpha(Faker is good): '+str(data.isalpha()))
print('9.2 isalpha(Fakerisgood): '+str('Fakerisgood'.isalpha()))
# 9.2. 숫자 검사 = isdigit
print('9.3 isdigit(sesco): '+str('sesco'.isdigit()))
print('9.4 isdigit(sesco121): '+str('sesco121'.isdigit()))
print('9.4 isdigit(121): '+str('121'.isdigit()))

# 10. 특정 문자열로 시작 또는 끝나는지 확인!
print('10.1 startswith(Faker): '+str(data.startswith('Faker')))
print('10.2 startswith(faker): '+str(data.startswith('faker')))
print('10.3 endswith(good): '+str(data.endswith('good')))
print('10.4 endswith(GOOD): '+str(data.endswith('GOOD')))
```
