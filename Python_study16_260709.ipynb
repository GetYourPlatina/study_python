```python
# 1. try-except 문

# try-except문은 try 블록 수행 중 오류가 발생하면 except 블록이 수행된다.
# 오류가 발생하지 않는다면 except 블록은 수행되지 않는다.
# 구문은 다음과 같다.

# try:
#    ...
# except [발생오류 [as 오류변수]]:
#    ...

# 여기서 [발생오류 [as 오류변수]] 는 생략할 수 있는 부분이다. []는 괄호 안의 내용을 생략할 수 있다는 관례적인 표기다.
# 발생오류를 적지 않으면, 모든 에러에 대해 except 문이 작동한다.
# 하지만 발생오류를 입력하면, 해당 발생오류에 대해서만 except 문이 동작하도록 구현할 수 있다.
# 여기에 오류변수까지 입력하면 발생오류의 내용이 저장되는 변수를 저장해준다.

# 실제 사용 예시
try:
    4/0
except:
    print('error is occurred')

print('')

# 실제 사용 예시
try:
    4/0
except ZeroDivisionError as e: # 발생오류가 있어야 오류변수 저장도 가능한듯...
    print(e)
```

```python
# try-finally 문
# try-finally 문은 오류가 발생해도 무조건 항상 수행되는 문이다.
# 주로 file을 열거나 했을때 오류가 발생하면 못닫는 경우가 생긴다. 
# 이때 이 finally문이 file을 닫을 수 있도록 아래와 같이 코딩할 수 있다.

try:
    f=open('test.txt','r')
    4/0
finally:
    f.close()
    print(f.closed)

# 아래 true를 보면 닫힌 것을 확인할 수 있다.
```

```python
# 여러 오류를 동시 처리하기
# 여러 오류를 동시 처리하되 각각 처리하는 방식

try:
    test_val=[1,2,3]
    print(test_val[3])
    4/0
except ZeroDivisionError as e:
    print(e)
except IndexError as e:
    print(e)

# 이떄 재미 있는 것은 가장 먼저 발생한 에러에서 except문이 수행되고 뒤의 index error는 수행되지 않는다.
# 아래와 같이 작성할 수도 있으니, 참고바람!
try:
    test_val=[1,2,3]
    print(test_val[3])
    4/0
except (ZeroDivisionError,IndexError) as e:
    print(e)
```

```python
# try-else 문
# try 문에는 다음과 같이 else 절을 사용할 수도 있다.
# 에러가 발생하지 않으면 else를 수행한다.
# 단 에러가 발생하면 except가 수행된다!

try:
    test_val=[1,2,3]
    print(test_val[1])
    4/1
except (ZeroDivisionError,IndexError) as e:
    print(e)
else:
    print('else go')
```

```python
# 오류 회피하기
# 오류가 발생해도 이를 무시하고 계속 진행하고 싶은 경우 except 문 아래 continue를 추가하면 된다.
score=[{'name':'Kim','score':50},{'name':'Lee'},{'name':'Hong','score':50}]

for x in range(len(score)):
    try:
        name=score[x]['name']
        num=score[x]['score']
        print(f'{name} 씨 점수는... {num} 입니다! 노력 ㄱㄱ')
    except:
        print(f'{name} 씨는 점수가 없네요... 시험을 안봤어?')
        continue

# 잘 동작하는 것을 확인할 수 있다.

# pass를 사용하는 방법도 있으나 pass는 신중하게 사용해야 한다. 
# 프로그램의 오류를 무시하다 중요한 오류까지 무시하면 나중에 더 큰 문제가 발생할 수도 있기 때문이다.
print('')

try:
    with open("init_set.txt",'r') as f:
        config=f.read()
except FileNotFoundError:
    pass # 설정파일 없어도 계속 진행
print('에러 있어도 계속 진행 무시 레고')
```
