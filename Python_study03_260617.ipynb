```python
# 튜플 자료형에 대해 알아보자
# 튜플은 리스트와 거의 비슷하다. 무엇이 다를까?
# 1. 리스트는 [], 튜플은 ()로 둘러싼다...
# 리스트는 요소값을 생성, 삭제, 수정할 수 있지만 튜플은 요소값을 바꿀 수 없다..
# 그럼 왜 튜플써요? - 프로그램 변수 중에서 변하지 말아야 하는 값(고정값)을 설정할 때는 도움이 될 수 있을 것이다.

print('tuple')
a=(1,2,3)
b=1,2,3
c=(1,)
d=(1) # 이거는 튜플이 아니다!

print(a)
print(b)
print(str(c) + ': it\'s tuple..') # 튜플을 단일로 생성하고 싶으면 (1,)로 생성할 것!
print(str(d) + ': it\'s not tuple..') # 튜플을 단일로 생성하고 싶으면 (1,)로 생성할 것! 이거는 튜플이 아니다!

# 튜플은 del이나 요소값을 변경할 수 없어서 앞서 list에 적용한 방법들이 통하지 않는다. 해보고 싶으면 아래의 코드를 실행해보라
# del a[1], a[1]=5

# 튜플 인덱싱 및 슬라이싱
# 앞서 다룬 list와 유사하므로 빠르게 지나간다.
a=(1,2,3,4,5,7,9,10,1,5,1)
print('slicing: '+str(a[5:]))
print('slicing:'+str(a[5:7]))
print('index(1):'+str(a[1]))

# 튜플 더하기, 곱하기
b=(11,55,44)
print('plus: '+str(a+b))
print('plus: '+str(a+(1,5)))
print('plus: '+str(a*2)) # 곱은 뭐.. 오직 정수만 ㅋㅋ

# 튜플 길이 구하기 - 똑같다.. len!
print('length a: '+str(len(a)))
print('length b: '+str(len(b)))


# 튜플 done..
```
