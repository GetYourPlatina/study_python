```python
# For 루프 기본 구조 살피기
# For루프는 while루프와 비슷하지만 반복회수를 정해준다는 것이 특징이다.
# For 문의 기본 구조는 다음과 같다.
print('normal for loop')
num=list(range(1,6))
for n in num:
    print(n)

# tuple이나 리스트를 for 문으로 쓸수도 있겠지!?
a=[(3,5),(7,9),(10,15)]
for (st,end) in a:
    print(str(list(range(st,end))))
    
pitcher=[['ryu',99],['moon',1],['kim',91]]
for [name,num] in pitcher:
    print('Today pitcher is No '+str(num)+', '+str(name))

# GOOOOOOOD
```

```python
# for 문을 활용하는 방법 예제
num=0
score=[55,44,88,99,100]

for n in score:
    num+=1
    if n > 50: print('%d 번째 학생은 점수가 %d 이므로 생존입니다.' % (num,n))
    else: print('%d 번째 학생은 점수가 %d 이므로 탈락입니다.' % (num,n))
```

```python
# for 문에서의 continue
# 동작은 while 문에서 사용한 것과 동일한 동작을 한다.
num=0
score=[55,44,88,99,100]

for n in score:
    num+=1
    if n <= 50: continue
    print('%d 번째 학생은 점수가 %d 이므로 생존입니다.' % (num,n))
```

```python
# range 함수와 활용의 예
# range 함수는 특정 구간의 숫자 범위를 만들어주는 함수다.
# range(시작,끝,간격)으로 설정한다.
# 값의 범위를 얻기 위해선 list와 함께 사용하여 list 자료형으로 변환을 해주는게 좋다. 안하면 그냥 range 객체로 남아있는다.
# 참고로 끝자리 수는 포함되지 않으므로 이를 감안하여 활용해야 한다.
a=range(0,9,1)
print(a)
print(list(a))
print('')

score=[55,44,88,99,100]
# 이를 활용해서 위의 예제를 다시 써보면 다음과 같다.
for n in range(len(score)):
    if score[n] <= 50: continue
    print('%d 번째 학생은 점수가 %d 이므로 생존입니다.' % (n+1,score[n]))
print('')

# range를 사용해서 간단한 구구단 만들기!
for i in range(1,5):
    for j in range(5,8):
        print(str(i*j),end=" ") #print의 끝(end는 \n(새로운줄)로 자동 설정되어 있는데, 이처럼 설정하면 띄어쓰기로도 바꿀수 있다.
    print('')
```

```python
# 리스트 컴프리헨션 문법
# list comprehension 직관적인 프로그램을 만들수 있는 문법으로 많은 부분에서 사용한다.
# 우선 아래와 같은 예제를 확인해보자
a=[1,2,3,4]
result=[]
for num in a:
    result.append(num*3)
print(result)

# 이를 컴프리헨션 문법으로 정리하면 아주 간단하게 작성이 가능하다.
# [표현식 for 항목 in 반복_가능_객체 if 조건문]
a=[1,2,3,4]
result=[num*3 for num in a]
print(result)

# for 문 안에 타 조건문(if) 등을 아래와 같이 추가할 수도 있다.
a=[1,2,3,4]
result=[num*3 for num in a if num%2==0] #2의 배수만 출력하도록 if 문 추가..
print(result)

# 컴프리헨션 문법에서도 다중 for문을 작성할 수 있다...
# [표현식 for 항목 in 반복_가능_객체1 if 조건문1
#        for 항목 in 반복_가능_객체2 if 조건문2
#        for 항목 in 반복_가능_객체3 if 조건문3
#        ...
#        for 항목 in 반복_가능_객체n  if 조건문n]

# 구구단 만들기!
result=[x*y for x in range(1,4)
       for y in range(1,5)]
print(result)

# 컴프리헨션으로 다중 if가 들어간 for문 만들어보기
score=[55,44,88,78,69]
result=['good 50<x<80: '+str(x) for x in score if x > 50 if x < 80]
print(result)
```

```python
# enumerate 함수 활용
# For의 경우 list를 입력하면 입력된 값이 차례대로 for문에 입력되는데 이때 인덱스 순서도 같이 들어갈 수 있게 만들수 있다.
# 바로 enumerate 함수를 활용하는 것이다.

pitchers = ['Ryu','Moon','Kim']
for i, name in enumerate(pitchers):
    print('%d번째 투수, ' % (i+1)+str(name))
print('')
# enumerate 함수는 시작점을 지정할 수 있다.
pitchers = ['Ryu','Moon','Kim']
for i, name in enumerate(pitchers,1):
    print('%d번째 투수, ' % (i)+str(name))
```

```python
# zip을 이용한 여러 리스트 활용 for 문 쓰기
# for 문을 활용해서 동작을 수행할 때, 하나가 아닌 여러 list를 동시에 입력시켜서 활용할 수 있다.
# zip을 이용하면 다수의 list를 동시에 입력하여 활용할 수 있다.

pitchers = ['Ryu','Moon','Kim']
numbers = [99,1,91]
score = [100,50,99]

for x,y,z in zip(pitchers,numbers,score):
    print('name: '+str(x)+', No.: '+str(y)+', score:'+str(z))


# For문 Done!!
```
