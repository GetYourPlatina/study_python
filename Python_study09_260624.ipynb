```python
# while 문의 기본 구조 살피기
# while 문은 문장을 계속적으로 반복해야될 때 사용하는 조건문이다.
# False가 나오지 않는다면 계속 반복된다. 무한으로 계속 즐기는 구조! (명XX사 갈비)


print('while loop example')
import time #소요시간 체크를 위함
loop_state=True
n=0

st_time=time.time()
while loop_state:
    n+=1
    if n==50:
        print('Rep 50')
        _50_time=time.time()
        print('%0.3f msec' % ((_50_time-st_time)*1e3))
    elif n==100:
        print('Rep 100 fin')
        _100_time=time.time()
        print('%0.3f msec' % ((_100_time-st_time)*1000))
        loop_state=False
    else:
        pass

# 입력을 받는 while문 예제를 만들어 보자
prompt="""
please input option number
1. lets go
2. setting
3. change menu
4. quit
"""

number=0
while number !=4:
    print(prompt)
    number=int(input())   
print('')
        
# while 문에서 강제로 빠져나가는 방법.. break
# while 문은 조건이 참(True)인 경우 계속 반복실행된다. 이를 강제로 종료시키는 방법이 바로 break 문이다.
# 이와 같은 break가 어떨때 필요할까? 자판기 예제로 한번 확인해보자
money=6000
coffee=5
while money>1000:
    money-=1000
    coffee-=1
    print('give cup of coffee after taking 1000 won.')
    if coffee==0:
        print('sold out')
        break
    else:
        print('Our machine have leftover %0.0f cup of coffee' % coffee)
    
# sold out이 되면 바로 break에 의해 바로 while 루프가 종료된다!
```

```python
# 자 그러면 자판기 한번 다시 만들어볼까? break를 활용해서 while 문 기반 자판기 만들기!

money=0
coffee=3 # 500원
water=2 # 300원

prompt="""
1. 돈 더 넣기
2. coffee
3. water
4. 돈 반환하기
"""

while True:
    if money>=300: 
        print('')
        print('메뉴를 골라주세요')
        state=int(input(prompt))
        match state:
            case 1: pass
            case 2:
                if (money>=500) & (coffee > 0): 
                    money-=500
                    coffee-=1
                    print('커피가 나옵니다. 남은 금액은 %0.1f 원 입니다.' % money)
                elif coffee <= 0:
                    print('커피는 품절되었습니다.')
                else:
                    print('돈이 부족합니다. 남은 금액은 %0.1f 원 입니다.' % money)
            case 3:
                if (money>=300) & (water > 0): 
                    money-=300
                    water-=1
                    print('물이 나옵니다. 남은 금액은 %0.1f 원 입니다.' % money)
                elif water <= 0:
                    print('물은 품절되었습니다.')
                else:
                    print('돈이 부족합니다. 남은 금액은 %0.1f 원 입니다.' % money)
            case _:
                print('남은 돈 %0.1f 원이 반환됩니다.' % money)
                money=0
        if (coffee==0) & (water==0): print('모두 품절되었습니다. 영업을 종료합니다.'); break
    else: 
        print('')
        input_money=int(input('돈을 넣어주세요: '))
        money=money+input_money
print('')
```

```python
# while 문의 맨 처음으로 돌아가기
# while 문 안의 조건문(if 등)을 수행할 때 입력 조건이 맞지 않으면 참(또는 거짓)값에 해당하는 동작을 스킵하고 이후 동작을 이어나간다. 
# 이때 이후 동작마저 모두 스킵하고 while 문의 맨 처음(조건문)으로 다시 돌아가게 만들고 싶은 경우가 있다. 
# 이때 사용하는 것이 바로 continue 문이다."

a = 0
while a<7:
    a+=1
    if a%5==0: continue; print(a)
    print(a)
# 아주 재미 있는 일이 생겼다! 위의 루프를 간단히 설명하자면
# 5의 배수일 경우 continue 문이 실행되도록 코드를 작성했다. 근데 continue 뒤의 print(a)도 실행되지 않았다!
# 즉 continue가 나오면 바로 while 루프의 시작점으로 돌아간다! 이거 잘 쓰일수 있다! ㄹㅇ
```

```python
# while 문에도... else가 있다!
# 다만 break를 쓰면 else를 통과하진 않는다. 
print('while else test')
count=0
while count<3:
    if count==2: print('Last!'); count+=1
    elif count==1: print('1 sec more'); count+=1
    else: print('moremore...'); count+=1
else: print('boom!')
print('')

print('while else test using break')
count=0
while count<3:
    if count==2: print('Last!'); count+=1; break
    elif count==1: print('1 sec more'); count+=1
    else: print('moremore...'); count+=1
else: print('boom!')
```

```python
# 중복 while 문...
# while 문을 중복해서도 사용할 수 있다. 

i=12
while i<=13:
    j=1
    while j<=9:
        k=i*j
        print('%d x %d = %d' % (i,j,k))
        j+=1
    i+=1
print('')

# break나 continue를 사용하면 가장 가까운! while 문에만 영향을 준다!
i=12
while i<=15:
    if i==14: break # 가장 가까운 while i<=15:에 영향을 준다.
    j=1
    while j<=5:
        if j==3: j+=1; continue # 가장 가까운 while j<=5:에 영향을 준다.
        k=i*j
        print('%d x %d = %d' % (i,j,k))
        j+=1
    i+=1

# while 문 Done!!
```
