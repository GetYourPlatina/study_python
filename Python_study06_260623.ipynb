```python
# Boolean for logical
# 불 자료형을 알아보자
# 우선 불(Bool)은 두가지 값만 가질수 있다. 참(True) 또는 거짓(False)

print('Bool')

# Bool 상수 및 논리연산 진행
a=True
b=False
print('  True value: '+str(a)+', class?: '+str(type(a)))
print('  False value: '+str(b)+', class?: '+str(type(b)))

c=5==5
d=5==6
print('  logical results?_(5==5): '+str(c))
print('  logical results?_(5==6): '+str(d))

e=5>=6
f=5>=5
print('  logical results?_(5>=6): '+str(e))
print('  logical results?_(5>=5): '+str(f))

# 자료형에서 참(True)과 거짓(False)
# 모든 자료형은 참과 거짓이 있다. 값이 있으면 참, 값이 비어있으면 거짓이다.
# 그밖에 숫자형은 0, 문자형은 ''이여도 거짓이며, None 상태도 거짓이니 참고하라.
# 이는 주로 제어문에서 활용하기 좋으니 이를 활용하는 예를 아래와 같이 참고해보자.

# 1: pop을 이용한 while loop 활용
print('example) how to use Data type boolean state')
state=[1,2,3,4]
print('state value: '+str(state))
while state:
    print(state.pop()) #pop은 기존 변수에서 하나씩 떼오는 기능, 불러온 값는 기존 변수에서 삭제됨. 기억하지?

# 2: 비어있는 리스트의 경우
if []:
    print('True')
else:
    print('False')

print('')

# Bool 함수 활용하기
# bool 함수를 쓰면 참인지 거짓인지를 알수 있음
py_good=[]
print('bool function use(empty list[]): '+str(bool(py_good)))

numfalse=0
numtrue=1
print('bool function use(0[]): '+str(bool(numfalse)))
print('bool function use(1[]): '+str(bool(numtrue)))

# Bool 자료형 설명은 우선 Done. 아래의 논리연산은 중요하니 잘 볼것
```

```python
# Bool 자료형의 논리연산(logical connective)
# Bool에는 and, or, not을 사용할 수 있다.
a=True
b=False

# and를 활용해보자!
print('operator and(&)')
print('T and T: '+str(a and a))
print('T and F: '+str(a and b))
print('F and T: '+str(b and a))
print('F and F: '+str(b and b))

print('T & T: '+str(a&a))
print('T & F: '+str(a&b))
print('')

# or를 활용해보자
print('operator or(|)')
print('T or T: '+str(a or a))
print('T or F: '+str(a or b))
print('F or T: '+str(b or a))
print('F or F: '+str(b or b))

print('T | T: '+str(a|a))
print('T | F: '+str(a|b))
print('F | F: '+str(b|b))
print('')

# not을 써볼까?(별도의 기호는 없는거 같다..)
print('operator not')
print('not(T): '+str(not a))
print('not(F): '+str(not b))
print('')

# XOR는 있는교? (That's right)
print('Exclusive OR(xor, ^)')
print('T ^ T: '+str(a ^ a))
print('T ^ F: '+str(a ^ b))
print('F ^ T: '+str(b ^ a))
print('F ^ F: '+str(b ^ b))

print('T and (not(T)): '+str((a and not(a)) or (not(a) and a)))
print('T and (not(F)): '+str((a and not(b)) or (not(a) and b)))
print('F and (not(T)): '+str((b and not(a)) or (not(b) and a)))
print('F and (not(F)): '+str((b and not(b) or not(b) and b)))

# NAND, NOR, XNOR는 not 게이트를 잘 사용하면 되므로 생략!

# 논리 연산자 Done
```
