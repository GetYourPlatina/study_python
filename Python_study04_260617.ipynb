```python
# 딕셔너리(dictionary)! - 연관배열(associative array)이라고도 한다.
# Key와 value가 한 쌍으로 대응되는 자료형이다. 
# 이름-Hkim, 나이-12, 직업-백수 와 같이 대응되는 자료형!

# 예시
print('Dictionary')
Hkim_info={'name':'HKim','age':'12','job':'white hand'}
print('1.1 Dictionary make: ' + str(Hkim_info))

# 정수와 문자열로도 딕셔너리를 만들수 있다.
Hkim_info={'name':'HKim','age':'12','job':'white hand',1:'point','good':2}
print('1.2 Dictionary make: ' + str(Hkim_info))

# 쌍 추가 및 삭제하기
# 3.7 버전 이전에는 쌍 추가 및 삽입 순서 등이 보장되지 않았지만 그 이후는 순서대로 저장되어 출력됨
Hkim_info[2]='is'
print('2.1 Add new key and value: ' + str(Hkim_info))
Hkim_info[0]='su'
print('2.2 Add new key and value: ' + str(Hkim_info))
Hkim_info['test']=['honaldo',1]
print('2.3 Add new key and value: ' + str(Hkim_info)) #딕셔너리의 value를 list로도 할 수 있다.
Hkim_info['test2']={'honaldo':1}
print('2.4 Add new key and value: ' + str(Hkim_info)) #딕셔너리의 value를 딕셔너리로도 할 수 있다.
# 징그러운 막 추가한 것들을 지워버리자
del Hkim_info['test']
print('2.5 Delete \'test\' value: ' + str(Hkim_info)) #딕셔너리의 value를 딕셔너리로도 할 수 있다.
del Hkim_info['test2']
del Hkim_info[2]
del Hkim_info[0]
print('2.6 Delete adding values: ' + str(Hkim_info)) #딕셔너리의 value를 딕셔너리로도 할 수 있다.
print('')

# 딕셔너리에서 key를 이용하여 값을 알아내기
# 딕셔너리는 값의 순서로 인덱싱을 하는 것과는 다르다! key를 입력해야 값이 나오는 것이므로 이를 유념할 것!
Team_name={'KHKim':'SK','HJRyu':'Hanhwa','DJMoon':'Hanhwa','SWPark':'Lotte'}
print('3.1 get value using key example dictionary: '+str(Team_name))
print('3.2 get KHkim team: '+str(Team_name['KHKim']))
print('3.3 get KHkim team: '+str(Team_name['SWPark']))
print('')

# 딕셔너리 만들때 주의사항
# 1. key는 중복된 값을 쓸수 없다.
Team_name={'KHKim':'SK','KHKim':'Hanhwa'}
print('4. caution example 1(Duplicate values): '+str(Team_name)) #앞서 넣은 값은 무시가 된다!
# 2. key는 list 값을 쓸수 없다. - Key는 값을 바꿀수 없고 list는 값을 바꿀수 있기 때문이다.
# Team_name={['KHKim','HJRyu']:'SK'}

# Dictionary Done!
```

```python
# Dictionary 관련 함수들 설명! 드간다!
Team_name={'KHKim':'SK','HJRyu':'Hanhwa','DJMoon':'Hanhwa','SWPark':'Lotte',1:'DWChoi'}
print('base dict.: ' + str(Team_name))
print('')

# 1. Keys - key 리스트를 만든다! - dict.에 포함된 key의 list를 만들어 준다! (하지만 list 자료형은 아닌...)
# dict_keys 객체를 반환하는 것으로 list와 사용은 유사하다. 다만 list의 고유 함수들(append)은 쓸수가 없다..
keys_name=Team_name.keys()
print('1.1 keys: ' + str(keys_name))
print(type(keys_name))
keys_name_list=list(Team_name.keys())
print('1.2 keys: ' + str(keys_name_list)) #list 함수를 쓰면 list로 변환도 가능하니! 참고
print(type(keys_name_list))
print('')

# 2. values - value의 리스트를 만드는 방법이다. - dict_values 객체를 반환하나 이도 역시 list를 쓰면 변환 가능!
values_name=Team_name.values()
print('2.1 values: ' + str(values_name))
print(type(values_name))
values_name_list=list(Team_name.values())
print('2.2 values: ' + str(values_name_list)) #list 함수를 쓰면 list로 변환도 가능하니! 참고
print(type(values_name_list))
print('')

# 3. items - 이건 key와 value를 튜플 자료형의 형태로 반환 받는 방법이다! 근데 객체는 dict_items인...
items_name=Team_name.items()
print('3.1 items: ' + str(items_name))
print(type(items_name))
items_name_list=list(Team_name.items())
print('3.2 items: ' + str(items_name)) #list 함수를 쓰면 list로 변환도 가능하니! 참고 ..?
print(type(items_name_list))
print('')

# 4. clear - 딕셔너리 내부를 모두 삭제함!
Team_name.clear()
print('4 clear: ' + str(Team_name))
Team_name={'KHKim':'SK','HJRyu':'Hanhwa','DJMoon':'Hanhwa','SWPark':'Lotte',1:'DWChoi'}
print('')

# 5. get - key 값에 맞는 value 찾기
# 이거는 사실 dict.에 key 입력하는 것과 동일(ex. Team_name['KHKim'] 하면 'SK'가 반환되는 것과 같아...) 그런데 key가 없을때 이 함수는 None을 발생시킴
# 그래서 key가 있는지 없는지 아리송 하면 get을 쓰는것도 추천!
print('5 get: ' + str(Team_name.get('KHKim')))
print('5 get: ' + str(Team_name['KHKim']))
print('5 get: ' + str(Team_name.get('KHKim1')))
print('')

# 6. in - key가 딕셔너리 안에 있는지를 조사하는 것으로 있으면 T 없으면 F의 불리언으로 출력이 나간다.
print('6 in(KHKim): ' + str('KHKim' in Team_name))
print('6 in(KHKim1): ' + str('KHKim1' in Team_name))

# 7. pop - get과 비슷하지만 pop은 값을 꺼내고 그 값이 있던 딕셔너리에서 반출된 값을 삭제하는 특징을 가진다.
print('7 pop(before): ' + str(Team_name))
print('7 pop(export): ' + str(Team_name.pop('KHKim')))
print('7 pop(after): ' + str(Team_name))


# Dictionary 함수들! Done!!
```
