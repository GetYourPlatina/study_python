```python
# 패키지(packages)
# 패키지란 관련 있는 모듈의 집합이며, 패키지를 사용하면 모듈을 계층적(디렉토리 구조)로 관리할 수 있음
# 파이썬에서 모듈은 하나의 .py 파일이다.
# 모듈이 많아지면 이름이 겹칠 수도 있고 관리가 어려워지는데, 패키지를 사용하면 모듈을 폴더별로 정리할 수 있어 많이 쓰인다.

# 아래는 패키지의 예시이다.

# /: directory, .py: module
# package_test_folder/
#    __init__.py
#    sound/
#       __init__.py
#       echo.py
#       wav.py
#    graphic/
#       __init__.py
#       screen.py
#       goal.py

# 위의 패키지를 직접 생성해본다!
# 참고로 각 디렉토리에 __init__.py 파일이 있어야 파이썬이 해당 디렉토리를 패키지로 인식한다. 
# 내용은 비워되 된다. 일단 생성해라!

# 다음으로는 터미널에서 패키지를 참조할 수 있도록 PYTHONPATH 환경 변수에 디렉토리를 추가한 다음 파이썬을 실행해보자
# export PYTHONPATH=/home/tester/work/python_study  - 우분투니까! 요거 씀!
# 패키지가 위치한 디렉토리를 정확하게 입력해야 호출이 가능하다.

# package_test_folder - sound directory에 있는 echo 모듈을 호출해보자!
# 터미널에서 아래와 같이 입력하여 불러오자
import package_test_folder.sound.echo
package_test_folder.sound.echo.echo_test()

# 위를 입력하면 결과가 잘 나오는 것을 확인할 수 있다.
# 참고로 터미널에서 파이썬을 실행하여 진행한 내용들은 반드시 'exit()'을 진행하여 import(가져온) 모듈, 객체 등등을 초기화할 필요가 있다.

# 불러오는 두번째 방법! from을 쓰는 것이다.
from package_test_folder.sound import echo
echo.echo_test()

# 세번째 방법은 아예 함수를 직접 import 하는 것이다.
from package_test_folder.sound.echo import echo_test
echo_test()

# 이때 import 주의사항! 도트로 불러올때는 반드시 마지막으로 불러오는 것은 모듈이나 함수를 import 해야 한다
# from package_test_folder import sound - 가능
# import package_test_folder.sound - 불가능

# 그리고 패키지만 불러오면 패키지 안의 함수인 __init__.py에 정의된 것만 참조할 수 있다.

# 그리고 함수까지 도트로는 불러올 수가 없다.
# import package_test_folder.sound.echo.echo_test 는 불가능
# echo_test는 echo에 종속된 함수일 뿐이지 패키지의 함수는 아니라서 그런듯..


# 헷갈리지 말기! 호출(call)은 함수를 실행하는 개념, 가져오기(import)는 객체, 함수, 클래스, 모듈 등을 쓰기위해 불러오는 것.

```

```python
# 그럼 __init__.py는 뭐에요?
# 앞서 이야기한 것과 같이 해당 디랙토리가 패키지의 일부임을 알려주는 역할을 한다.
# python3.3 이후론는 없어도 인식하나... 관례상 만들어주자
# init이 initalize의 줄임말인듯? 즉 패키지의 설정이나 초기화 코드를 포함시키기도 한다고 한다요

# import를 할때 패키지만 불러오면 패키지 안 __init__.py에 정의된 함수만 참조 가능합니다.
# 이를 터미널에서 불러오면 해당 함수는 잘 동작하는 걸 알수 있습니다.
import package_test_folder
package_test_folder.version_print()

# 자 이점을 이용하면 init에 각 모듈이나 함수를 미리 import 시켜놓으면 매우매우 편하게 활용 가능하다는점!
# 잘 활용하시길 바랍니다.
# 아래의 디렉토리와 함수들을 불러왔다. 한번 임포트 해서 불러와보자!
# from package_test_folder import sound
# from package_test_folder.sound.echo import echo_test
# import package_test_folder.graphic.goal

# 우선 실습전에 커널(또는? interpreter?) 초기화 부터!
exit()
```

```python
# __all__에 대해 알아보자
# 패키지를 불러올 때, 모듈 내 모든 함수를 불러오려면? *을 쓴다.
from package_test_folder.sound.echo import *
print(dir())

print('')
exit()
```
