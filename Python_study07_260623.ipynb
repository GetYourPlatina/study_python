```python
# Variable 변수
# 변수는 변할 수 있는 수! 즉 사용자가 임의로 변화를 줄수도 있고 연산으로 변화가 가능한 수이다.
# C와 java에서는 변수를 만들때 자료형을 직접 지정해야 되지만, 파이썬은 그렇지 않다. 자동으로 자료형을 지정하기 때문에 매우 편하다.

# 변수명의 규칙! 이것은 명심할 것!
# 1. 영문자, 숫자, 언더스코어(_)만 사용가능하다.
# 2. 숫자로 시작할 수 없다.
# 3. 예약어(파이썬에서 특별한 의미(함수명이나 특정 값으로 쓰이는 것 등)로 사용되는 예약어들은 변수로 사용 불가
# 4. 대소문자를 구분짓는다. 

# 변수명의 권장사항
# 1. 의미가 명확한 이름을 사용한다.
# 2. snake_case(단어와 단어 사이에는 언더스코어를 권장)
# 3. 너무 길거나 짧은 이름은 피한다.

# 변수명 예시
Name='버러범범'
Age=40
user_Name='GYP'
_private='secret'

# 변수는 객체를 가르키는 것과 같다. Age=40 에서 'Age'가 변수이다. 40은 해당 변수에 대응되는 객체(데이터)라 보면 된다.
# Age=40을 실행하면 객체가 40인 변수 Age가 메모리에 생성되어 저장된다. 이때 저장된 메모리의 주소는 다음과 같이 찾을수 있다.
Age_address=id(Age)
print("variable address(Age_original.): "+str(Age_address))

# 재미 있는 것은 변수 Age를 복제하여 id를 확인해보면 주소는 동일한 것을 알 수 있다.
Age_copy=Age
Age_copy_clone=id(Age_copy)
print("variable address(Age_copy.....): "+str(Age_copy_clone))
# 즉! 변수를 copy 한다고 해서 새로운 객체가 생성되어 저장되는 것이 아니다. 동일한 객체면 해당 객체에 대응되는 변수가 늘어나는 것 뿐이지 객체가 늘어나서 새로 메모리에 저장(할당)되는 것은 아니다.
# 자 그렇다면? 동일한 객체에 다른 변수를 지정하면 어떻게 될까? 확인해보자.
goodi=40
goodi_clone=id(goodi)
print("variable address(same_object..): "+str(goodi_clone))
# wow! 같지 않은가? 즉 메모리에 할당되는 데이터는 객체이고 변수는 그 객체를 가르키는 꼬리표? 같은 것이다. 동일한 객체에 여러 변수가 붙을수 있다. 
# 즉 변수는 달라도 객체가 같다면? 메모리 address가 같다!
print('')

# 움... list도 그래요? 한번 확인해보자
# a와b는 카피를 하고, c는 동일한 객체의 리스트를 생성해본다.
a=['ryu',99]
b=a
c=['ryu',99]
print("variable address(a list): "+str(id(a)))
print("variable address(b list): "+str(id(b)))
print("variable address(c list): "+str(id(c)))

# 엇? 재미있는 일이 생겼다. a와 b는 카피로 동일한 객체에 변수임이 확인되었다.
# 그런데? c는? a와 b와 구조 및 값이 같아도 다른 객체임을 확인할 수 있다.
# 즉! 리스트는 copy를 통해서만 같은 객체에 다른 변수를 할당할 수 있다.
# 같은 구조와 값의 리스트를 생성한다고 기 생성된 객체에 다른 변수를 할당하는 것은 아니다.
# 더더더더 재미있는 것을 보여주겠다.

a.append(77)
# 위와 같이 77을 생성해주면 b는 어떻게 될까? a가 다른 메모리에 객체를 생성하는 것일까? 아니면 b도 동일하게 객체가 변경되는 것일까?
print("a object: "+str(a)+", variable address(a list)_modified: "+str(id(a)))
print("b object: "+str(b)+", variable address(b list)_modified: "+str(id(b)))
# 놀랍게도 b의 list도 a와 같이 변경되고 address는 동일한 것을 확인할 수 있다. 
# 즉 여기서 중요한 점은 객체 위주로 데이터가 운용되는 것이다. 변수에 따라서가 아니다!
print('')

# copy를 하는 것이 단지 객체에 새로운 변수를 추가하는 일이면 문제가 발생할 수 있다.
# 객체를 복사하여 복사 전 원본이 되는 객체는 유지하되 복사한 객체는 수정하여 활용하고 싶을 때가 많을 것이다. 그런 경우 카피는 어떻게 하면 좋을까?
# 방법이 있다! 바로 [:]를 이용하는 것이다.
print('- object copy method')
a=['ryu',99]
b=a
c=a[:]
print("a object: "+str(a)+", variable address(a list): "+str(id(a)))
print("b object: "+str(b)+", variable address(b list): "+str(id(b)))
print("c object: "+str(c)+", variable address(c list): "+str(id(c)))

# 다른 방법은 copy 모듈을 활용하는 법이다.
from copy import copy
d=copy(a)
print("d object: "+str(d)+", variable address(d list): "+str(id(d)))
print('')

# 서로 객체가 같은지 다른지 확인해보자!
print('- object address check')
print('a=b?: '+str(a is b))
print('a=c?: '+str(a is c))
print('a=d?: '+str(a is d))
print('')


# gooooood! 좋다! 마지막으로 변수를 만드는 여러가지 방법을 알아보도록 하자
# 변수를 만드는 여러가지 방법! 튜플이나 리스트로도 변수를 만들수 있다.
a,b=('ryu',99)
print('a object: '+str(a)+', a type: '+str(type(a)))
print('b object: '+str(b)+', b type: '+str(type(b)))

(a,b)='ryu',99
print('a object: '+str(a)+', a type: '+str(type(a)))
print('b object: '+str(b)+', b type: '+str(type(b)))

[a,b]=('ryu',99)
print('a object: '+str(a)+', a type: '+str(type(a)))
print('b object: '+str(b)+', b type: '+str(type(b)))

# [a,b]='ryu' 이건 에러가 발생! 2개의 변수가 있다면 거기에 맞게 객체도 지정해줘야 한다.
[a,b]=('ryu',[a,b]) # 요건 된다!
print('a object: '+str(a)+', a type: '+str(type(a)))
print('b object: '+str(b)+', b type: '+str(type(b)))

# a와 b를 다음의 방법으로 교체할 수도 있다.
(a,b)='ryu',99
a,b=b,a
print('a object: '+str(a)+', a type: '+str(type(a)))
print('b object: '+str(b)+', b type: '+str(type(b)))

#객체 Done
```
