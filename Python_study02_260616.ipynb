```python
# List 자료형
# 문자형과 숫자형은 단순히 하나의 데이터를 저장하고 사용했다. 그러면 숫자 꾸러미나 문자의 꾸러미는 어떻게 변수로 저장하고 활용할수 있을까>
# 이번에는 리스트에 대해 알아보도록 하자!
# 1. 다양한 리스트 만들어보기!
num_list=[1,5,7,4,6]
str_list=['apple','banana','yummmm']
numstr_list=[1,2,'good','faker']
multi_list=[1,2,'good',[3,4,'sexy']]
super_multi_list=[1,'Moon',[99,'Ryu'],num_list]
print('1. list data examples:')
print(num_list)
print(str_list)
print(numstr_list)
print(multi_list)
print(super_multi_list)
# 비어있는 리스트는 아래의 함수로 생성 가능하다.
null_list=list()
print(null_list)
print('')

# 2. 리스트의 인덱싱과 슬라이싱
# 슬라이싱은 0<=ㅌ<1 임을 잊지 마라!
print('2. list indexing and slicing')
print('  indexing(num_list_1): '+str(num_list[1]))
print('  indexing(multi_list_3): '+str(multi_list[3]))
print('  indexing(multi_list_3): '+str(multi_list[-1])) # 역방향 인덱싱도 가능!
print('  indexing(multi_list_3 and 0): '+str(multi_list[3][0])) # 리스트 안 리스트를 인덱싱 하려면 이 방법으로 가능!
print('  slicing(num_list_234): '+str(num_list[2:])) # 2위치를 포함하고 슬라이싱!
print('  slicing(num_list_23): '+str(num_list[2:4])) # 2-4위치 슬라이싱
print('  slicing(multi_list_3): '+str(multi_list[2:])) # 변수,리스트도 슬라이싱이 가능
print('  slicing(multi_list_3): '+str(multi_list[2:][1])) # 변수,리스트를 슬라이싱 한것에서 인덱싱 하는 것도 가능
print('  slicing(multi_list_3): '+str(multi_list[2:][1][:2])) # 인덱싱한 것에서 다시 슬라이싱도 가능!
print('')

# 3. 리스트 연산하기
# 리스트를 연산할 수 있는데 수식을 계산하는 그런건 아니고 걍 추가하거나 반복하거나 하는것이니 참고
print('2. operating list')
print('  operating list(+): '+str(num_list+num_list))
print('  operating list(*3): '+str(num_list*3))
print('')

# 4. 리스트 길이 구하기
# len 쓰면 된다..
print('3. list length(len)')
print('  len(num_list): '+str(len(num_list)))
print('  len(multi_list): '+str(len(multi_list)))
print('')

# 5. 리스트 수정 및 삭제
# 리스트는 수정과 삭제가 가능하다!
# num_list=[1,5,7,4,6]
# str_list=['apple','banana','yummmm']
# numstr_list=[1,2,'good','faker']
print('4. list revise and remove')
num_list[1]=55
multi_list[3]=55
print('  revise: '+str(num_list))
print('  revise: '+str(multi_list))
multi_list[3]=[99,'Ryu']
print('  revise: '+str(multi_list))
# 다만 삭제는 del 함수를 써야 한다..
del num_list[1]
del multi_list[3]
print('  remove 55(location 1): '+str(num_list))
print('  remove [99,\'Ryu\'](location 3): '+str(multi_list))
# slicing을 활용할 수도 있다.
del num_list[1:]
print('  remove with slicing(location 1:): '+str(num_list))
# 초기화..
num_list=[1,5,7,4,6]
numstr_list=[1,2,'good','faker']
print('')

# 리스트 기본 설명 done
```

```python
# List 관련 함수들 설명!
multi_list=[1,2,'good',[3,4,'sexy']]
print('Base list: '+str(multi_list))
print('')

# 1. append - 리스트에 요소 추가하기, 리스트의 가장 마지막 자리에 요소를 추가하게 됨
multi_list.append([99,'Ryu'])
print('1. append([99,Ryu]): '+str(multi_list))
del multi_list[4]
print('')

# 2. sort - 리스트를 정렬해준다..
num_list=[1,4,5,2,3]
str_list=['b','c','a']
multi_list=[2,'good',1]
num_list.sort()
str_list.sort()
# multi_list.sort() 숫자 문자가 복합된 list는 지원하지 않음!

print('2. sort(num): '+str(num_list)) #sort까지 안으로 가져오면 none으로 뜸... 사유는 모름
print('2. sort(num_test..): '+str(num_list.sort())) #sort까지 안으로 가져오면 none으로 뜸... 사유는 모름
print('2. sort(str): '+str(str_list))
# print('2. sort(str): '+str(multi_list)) 숫자 문자가 복합된 list는 지원하지 않음!
# 초기화
multi_list=[1,2,'good',[3,4,'sexy']]
print('')

# 3. reverse - 리스트 뒤집기
multi_list.reverse()
print('3. reverse: '+str(multi_list))
# 초기화
multi_list=[1,2,'good',[3,4,'sexy']]
print('')

# 4. index - 인덱스 반환: 찾고자 하는 요소를 넣으면 위치를 알려줌
index_list=multi_list.index('good')
print('4. index(\'good\'): '+str(index_list))
index_list=multi_list.index([3,4,'sexy'])
print('4. index([3,4,\'sexy\']): '+str(index_list))
#index_list=multi_list.index(4) 리스트 안에 리스트 항목은 찾을수 없다.
print('')

# 5. insert - 리스트에 요소 삽입!: 첫번쨰는 위치, 두번째는 값!
multi_list.insert(0,47)
print('5. insert(0,47): '+str(multi_list))
multi_list.insert(8,[47,'kwon']) # 위치를 저멀리 둔다한들 지정 리스트의 가장 마지막 위치로 강제되어 입력되니 참고!
print('5. insert(0,47): '+str(multi_list))
# 초기화
multi_list=[1,2,'good',[3,4,'sexy']]
print('')

# 6. remove - 리스트 요소 제거!:지우고자 하는 요소를 넣으면 위치를 알려줌
multi_list.remove('good')
print('6. remove(\'good\'): '+str(multi_list))
# 초기화
multi_list=[1,2,'good',[3,4,'sexy']]
print('')

# 7. pop - 요소 꺼내기. 꺼내면 꺼내진 요소는 list에서 삭제됨, 이건 위치를 지정이야
tt=multi_list.pop()
print('7. pop(): '+str(multi_list))
print('7. pop(export): '+str(tt))
# 초기화
multi_list=[1,2,'good',[3,4,'sexy']]
tt=multi_list.pop(2)
print('7. pop(2): '+str(multi_list))
print('7. pop(export): '+str(tt))
# 초기화2
multi_list=[1,2,'good',[3,4,'sexy']]
print('')

# 8. count - 요소 세기! - 안에 찾는 요소의 갯수를 센다!
test_list=[1,2,3,4,5,5,5,5,7,1,1]
print('8. count example: '+str(test_list))
print('8.count(1): '+str(test_list.count(1)))
print('8.count(5): '+str(test_list.count(5)))
print('')

# 9. extend - 리스트 확장! 요거는 리스트만 확장할 수 있음!
a=[1,2,3,4,1]
#ex_test_list=test_list.extend(a) 요런 방식 불가능 오직 리스트 직접 입력만..
test_list.extend([2,3,4])
print('9. extend: '+str(test_list))
test_list+=[2,3,4]
print('9. extend: '+str(test_list))
test_list=test_list+[2,3,4]
print('9. extend: '+str(test_list))
# 요 방식들 다 동일하다...

# list 관련 함수들 Done!!
```
