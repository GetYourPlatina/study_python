```python
# if 문(if statement) 
# if 문은 주어진 상황에 맞게 행동을 처리할 수 있도록 하는 기능이다.
# 말로 설명해보면 다음과 같을 수 있다.
#  - 돈이 만약 5000원 이상을 가지고 있다면 택시를 탄다. 아니면 걸어간다.
#  - if(조건): 돈 >= 5000원 
#  - 조건이 충족할 때 행동(Do): take a taxi
#  - 조건 불충족(else): just... walk

# 이를 if 문으로 나타내보면 다음과 같다.
print('if statement ex 1.')
oney=5000
if money >= 5000:
    print('take a taxi')
else:
    print('walk booooy')
print('')

print('if statement ex 2.')
money=2000
if money >= 5000:
    print('take a taxi')
else:
    print('walk booooy')
print('')

# 자 여기서 중요한 부분은 들여쓰기를 잘 지켜야 한다는 것이다. if 문 안에 행동은 철저하게 들여쓰기로 작성해야 한다.
# 아래와 같이 작성하면 들여쓰기가 잘못되어 에러가 출력된다...
print('if statement ex 3.')
# if money >= 5000:
# print('take a taxi')
# else:
#    print('walk booooy')
# 들여쓰기가 일치하지 않으면 에러가 발생한다.
# print('')
# money=20200
# if money >= 5000:
#     print('take a taxi')
#         print('tt')  # 이렇게 들여쓰기가 불일치하면 에러가 생겨
# else:
#     print('walk booooy')
# 들여쓰기가 많이 들어가도 일치만 하면 에러 없이 실행은 된다.
money=20200
if money >= 5000:
        print('take a taxi')
        print('Yessss')  # 이렇게 들여쓰기가 불일치하면 에러가 생겨
else:
    print('walk booooy')
print('')
# 들여쓰기는 탭 또는 공백을 쓰는데 이 둘을 써도는 된다. 하지만 코드 내에서 일관되게 탭이면 탭, 공백이면 공백으로... 통일하자
# 파이썬 커뮤에서는 공백 4번이 거의 통용된다고는 하는데.... 모르겠네.. ㅋㅋ
# 조건문에는 반드시 콜론(:)을 붙여야 한다. 다른 언어들은 {}로 감싸지만 요건 아니다..

# 조건문에 대해서 알아보자! x (조건문) y의 형태로 쓰이는데 x는 비교대상 , y는 기준치가 된다.
# x < y: x가 y보다 작은가?
# x <= y: x가 y보다 작거나 같은가?
# x > y: x가 y보다 큰가?
# x >= y: x가 y보다 큰거나 같은가?
# x == y: x와 y가 같은가?
# x != y: x와 y가 같지 않은가?

# 해당 조건들이 충족하면 True를 충족하지 않으면 False를 반환한다.
# 쉬운 내용이니 구지 예제 없이 가자!

# 조건문에는 and, or, not 등의 논리 연ㅅ나자도 활용 가능하다. 
# 관련 내용은 앞서 연습한 bool 자료형을 참고하자... True, False가 어떻게 되는지 확인가능하다.
# 이를 활용한 예제를 아래와 같이 남겨본다.
print('if statement ex 4(with logical operation).')
money=5000
money_in_card=3000
if money >= 6000 or money_in_card >= 6000 or money+money_in_card >= 6000:
    print('you can hang out with your girlfriend.. haha!!')
else:
    print('you just should be working hard... loser...TT')
print('')

# in, not in - 파이썬에만 있는 재미있는? 조건문이다... 보통 list를 활용한 if 문으로 보여지는데
# list 안에 찾고자 하는 요소가 있는지? 없는지?를 판단하고 True 또는 False를 낸다.
print('conditional statement(in)')
a=[1,2,'Ryu',99]
Ryu_in_here='Ryu' in a
print('  list a is: '+str(a))
print('  \'Ryu\' in list(a)?: '+str(Ryu_in_here))
New_not_in_here='New' not in a
print('  \'New\' not in list(a)?: '+str(New_not_in_here))
print('')

# in을 활용한 if문은 아래와 같을것이다.
print('if statement ex 5(with in or not in).')
pitcher_list=['Ryu','Moon','Kim']
if 'Ryu' in pitcher_list:
    print('This game will be win! broooo...')
    print('start pitcher is normally '+str(pitcher_list[0]))
    print('closing pitcher is normally '+str(pitcher_list[2]))
else:
    print('This game will be lose...')
print('')


# 조건문에서 조건이 맞을때 아무 일도 일어나게 하지 않게 하려면? pass를 쓴다! 위의 예제를 그대로 적용해보자.
print('if statement ex 6(with pass).')
pitcher_list=['Ryu','Moon','Kim']
if 'Ryu' in pitcher_list:
    pass
else:
    print('This game will be lose...')
print('nothing is display.')

# if문은 한줄로도 작성할 수 있다. 위의 코드를 한줄로 적어보면
pitcher_list=['Ryu','Moon','Kim']
if 'Ryu' in pitcher_list: pass
else: print('This game will be lose...')
print('nothing is display.')
# 동작은 똑같다. 콜론(:) 뒤에 내용을 넣어도 된다.

# if와 관련된 일반적인 내용 done.
```

```python
# if 관련 조건문 심화!
# 여러 조건을 담은 조건문을 작성할 때
money=5000
card=3000

if money > 6000:
    print('take a taxi')
else:
    if card > 2000:
        print('take a subway')
    else:
        print('walk... bro')
# 위와 같이 작성할 수 있다... 근데 복잡하지 않은가? 그래서 준비했다!

# 다중 조건문을 만들때! elif
# elif는 상위 if나 elif에서 거짓이 발생했을때 수행되는 다음의 조건이다. elif를 위의 코드에 적용하면 다음과 같다.
if money > 6000:
    print('take a taxi')
elif card > 2000:
    print('take a subway')
else:
    print('walk... bro')
# 훠어어어얼씬 깔끔하다. 그리고 코드가 직관적으로 보인다. 이를 적극 활용하라!
print('')


# match-case 문!
# python 3.10 이후 버전에서 사용할 수 있는 조건문이다. 특정 조건일 때 행동을 수행하도록 조건문을 생성할 때 쓰면 좋다.
# 예시를 들어보자.
grade='b'

match grade:
    case 'a':
        print('wow... you genius')
    case 'b':
        print('Normal')
    case 'c':
        print('great!')
    case _:
        print('get out')
# 언더바는 지정된 case를 제외한 모든 값의 경우를 말한다. 이를 생략하면 상위 case가 없는 경우엔 값이 return 되지 않는다.
# 이를 안쓰면 어떻게 해야 되었는가?
if grade == 'a': print('wow... you genius')
elif grade == 'b': print('Normal')
elif grade == 'c': print('grate')
else: print('get out')      
# 짧긴 하지만...가독성이 좀 떨어진다 볼수도..
# 뭐 여튼 match-case 문도 좋은 옵션이 될 수 있다.

# 그리고 각 case 조건에 or(|)를 활용할 수도 있으니 참고하라
match grade:
    case 'a'|'b'|'c':
        print('wow... you genius')
    case _:
        print('get out')
print('')

# 연산자 연쇄사용.. 대략 비교연산자를 연속으로 쓰는 내용이며 아래와 같으니 간단히 참고
a=10
check= 2<a<4
check2= 8<a<=11
print(check); print(check2); print('');

# 조건부 표현식(생각보다 많이 쓰니까 잘 알아두도록 해라!!)
# 복잡하지 않은 단순 분기 조건문일 떄 주로 사용한다. 아래와 같은 조건문이다.
score=50
if 0<score<=50:
    state='re-test lets go'
else:
    state='pass!'
print(state)
# 위와 같은 조건문을 아래와 같이 조건부 표현식으로 간단하게 표현이 가능하다.
score=85
state= 're-test lets go' if 0<score<=50 else 'pass!'
print(state)

pitcher_list=['Ryu','Moon','Kim']
results_game='win' if 'Ryu' in pitcher_list else 'lose..'
print(results_game)
# 하지만 복잡한 조건 등이 붙는 경우에는 그냥 if-elif 등을 잘 활용해보자!

# if 문 done!!
```
