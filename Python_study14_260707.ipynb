```python
# 모듈에 대해서 알아보자
# 모듈이란 함수나 변수 또는 클래스를 모아 놓은 파이썬 파일을 의미한다.
# 다른 파이썬 프로그램에서 불러와 사용할 수 있게 만든 파이썬 파일이라고도 할 수 있다.
# 모듈을 불러오기 위해선 python 실행 시 설정된 파일 경로 상에 파이썬 파일이 존재해야 한다.
# 해당 예제에서 파일 경로는 다음과 같았다.
# '/home/tester/work/python_study'
# 경로를 확인하는 법은 os 모듈을 쓰는 법이다. (os는 표준모듈(built-in module)이라 경로 문제 없이 import가 가능하다)
import os
os.getcwd() # getcwd method를 쓰면 경로 확인이 가능해진다.

# 해당 파일에 module_test1.py 를 아래와 같이 생성했다.
# def add(a,b):
#     return a+b
# def sub(a,b):
#     return a-b

# 한번 import를 해보자!
import module_test1
# import가 잘 된 것을 확인할 수 있다. 자 이제 사용해볼까?
result1=module_test1.add(10,20)
print(result1)
result2=module_test1.sub(10,20)
print(result2)
# 내부의 메서드 역시 잘 동작함을 알 수 있다.

# 위와 같이 쓰는 것이 좀 불편할 수 있다. 모듈명을 작성하고 점을 작성하고 또 함수(메서드)를 꺼내와야하는 불편함..
# 이때 사용하는 것이 from이다. from을 사용해서 모듈 내 함수를 직접 호출할 수도 있다.
from module_test1 import add
from module_test1 import sub
result3=add(10,20)
result4=sub(10,20)

print(f'result 1: {result1} and result 3: {result3}, same!')
print(f'result 2: {result2} and result 4: {result4}, same!')
# 값을 비교해봐도 잘 동작하는 것을 알수 있다.

#함수 add와 sub를 위에선 하나씩 불러왔지만 아래와 같이 한번에 불러올 수도 있다.
from module_test1 import add,sub # 콤마로 복수의 함수를 불러올 수도 있고
# or
from module_test1 import * # *을 써서 내부의 모든 함수를 불러올 수도 있다.
```
