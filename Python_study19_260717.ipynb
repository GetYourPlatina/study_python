```python
# 표준 라이브러리 가즈아!

# operation.itemgetter
# sortting에서 조건을 부여하는 것으로, 딕셔너리 자료형이나 복합 튜플,리스트 형과 같은 다양한 데이터가 있는 경우 정렬 위치를 지정할 수 있게 해주는 함수이다.

from operator import itemgetter

student=[['ryu',55,'a'],['moon',10,'v'],['ssi',30,'b']]

test1=sorted(student, key=itemgetter(1))
print(test1)

# 딕셔너리일때는 key를 직접 입력하면 된다!

students=[{'name':'kim','age':54,'grade':'c'},
         {'name':'ryu','age':11,'grade':'a'},
         {'name':'moon','age':5,'grade':'b'}]

test2=sorted(students, key=itemgetter('age'))
print(test2)



# 단 클래스의 객체를 정렬하려면 itemgetter가 아닌 attrgetter()를 서야 한다.
from operator import attrgetter

class student:
    def __init__(self,name,age,grade):
        self.name=name
        self.age=age
        self.grade=grade

students=[
    student('ryu',44,'a'),
    student('moon',1,'d'),
    student('kim',52,'c')]

test3=sorted(students,key=attrgetter('age'))
print(test3[0].name,test3[1].name,test3[2].name)
```
