```python
# 집합(set) 자료형에 대해서 알아보자
# 집합형은 수학의 집합 개념과 동일한 자료형임. 중복이 허용되지 않고! 순서가 없는 데이터 모임으로, 교집합, 합집합, 차집합 등의 연산이 용이함
# How to make set?
# set을 이용하여 만들거나
print('set')
print('1. How to make set?: ')

num_set=set([1,2,3])
str_set=set("hello")
print('  1.1 num set: ' + str(num_set))
print('  1.2 str set: ' + str(str_set))

num_set2={1,2,3}
str_set2={'a','b','c'}
print('  1.3 num set using curly braces{}: ' + str(num_set2))
print('  1.4 str set using curly braces{}: ' + str(str_set2))

# 참고로 딕셔너리도 중괄호인 {}를 쓰다보니 비어있는 집합을 만드는 법은 아래와 같다.
none_set=str()
# none_set={}는 비어있는 딕셔너리를 만드니 참고하라


# 집합은 앞서 이야기 했듯이 중복이 허용되지 않고 순서가 없다!
# 이 때문에 인덱싱이 되지 않는다. 인덱싱을 하고 싶다면, 리스트나 튜플로 변환하면 된다.
str_tuple=tuple(str_set2)
str_list=list(str_set2)
print('  2.1 set indexing(convert set 2 tuple): ' + str(str_tuple))
print('  index 2: '+str_tuple[2])
print('  2.2 set indexing(convert set 2 list): ' + str(str_list))
print('  index 2: '+str_list[2])


# 앞서 이야기 한 것과 같이 집합은 교집합 합집합 차집합을 구할때 아주 편하다.
print('2. intersection, union, difference: ')
S1={1,2,3,5,7,11}
S2={1,2,4,8,16,32}
print('  S1: '+str(S1))
print('  S2: '+str(S2))
print('')

# 교집합, &
intersection_s=S1&S2
intersection_s2=S1.intersection(S2)
print('  intersection: '+str(intersection_s)+'  '+str(intersection_s2))

# 합집합, |
union_s=S1|S2
union_s2=S1.union(S2)
print('  intersection: '+str(union_s)+'  '+str(union_s2))

# 차집합, |
difference_s=S1-S2
difference_s2=S1.difference(S2)
difference_s3=S2-S1
difference_s4=S2.difference(S1)
print('  intersection: '+str(difference_s)+'  '+str(difference_s2))
print('  intersection: '+str(difference_s3)+'  '+str(difference_s4))

# Done
```

```python
# 집합(set) 자료형 관련 함수들
S1={1,2,3,5,7,11}
S2={1,2,4,8,16,32}

# 1. add - 값 1개 추가하기
S1.add(19)
print('  1.1 add(num): '+str(S1))
# 이미 있는 요소를 추가하면 추가되진 않아... 중복은 안되기 때문..
S1.add('tt')
print('  1.2 add(str): '+str(S1))

# 2. update - 값 여러개 추가하기
S2.update([5,4,9])
print('  2.1 update(str): '+str(S2))
S2.update([99,77,'Ryu'])
print('  2.2 update(complex): '+str(S2))

# 3. discard - 특정 값 제거하기 (1개..)
# list는 입력이 안되고, 다른 set이나 tuple은 삭제가 안됨....
S2.discard('Ryu')
print('  3.1 discard(str): '+str(S2))
S2.discard(99)
print('  3.2 discard(num): '+str(S2))
S2.discard('good')
print('  3.2 discard(nothing): '+str(S2))
# 없는 것을 지우려고 해도 삭제되지 않는다.

# 4. clear - 모든 값 제거하기
S1.clear()
print('  4 clear: '+str(S1))

#set의 기능들 done
```
