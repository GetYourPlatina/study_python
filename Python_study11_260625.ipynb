```python
# Function(함수)
# 함수란 입력(x) 대비 특정 동작을 진행한 뒤 결과를 출력(y) 하는 역할을 하는 것이다.
# ex) y=2x+3도 함수이다.
#
# 파이썬에서 함수의 구조는 아래와 같다.

# def 함수이름(매개변수):
#    수행할_기능
#    수행할_기능 2

# + 역할을 하는 함수를 아래와 같이 만들어보면?
print('1. function example. plus function making')
def plus_exam(x,y): #여기서 x,y는 매개변수
    return x+y

result=plus_exam(10,5) #여기서 10,5는 인수
print(result)
print('')
# 결과를 보면 잘 동작하는 것을 알 수 있다.

# 매개변수(parameter)와 인수(arguments)
# 매개변수는 함수에 입력으로 전달된 값을 받는 변수
# 인수는 함수를 호출할 때 전달하는 입력값
# 말로 설명하는 것보다 보는 것이 더 이해가 빠를 것이다.
# 위의 함수를 예로 들면
# 매개변수는 def plus_exam(x,y)에서 x,y를 의미하며
# 인수는 plus_exam(10,5)와 같이 실제로 사용할 때 입력된 값을 의미한다.

# 용어 통일(점프 투 파이썬과 동일하게)
# 입력값 -> 함수 -> 반환값(리턴값) 으로 통일하자

# 입력값이 없는 함수 생성 - 입력값이 없는 함수가 있을까? 한번 확인해보자
print('2. function example. no input')
def say():
    return 'hello world'
    
a=say()
print(a)
print('')
# 위 처럼 입력값이 없는 함수도 만들 수 있다. 나는 보통 초기화 함수를 생성할 때 요런식으로 쓰긴 한다만.. 여튼 뭐 꽤나 많이 쓴다.

# 반환 값이 없는 함수 생성 - 반환값이 없는 함수도 있을까? 한번 확인해보자
print('3. function example. no return')
def say(x):
    print(x+' hello world')
    
a=say('ryu')
print(a)
print('')
# 단 리턴이 없다보니 a 값이 None으로 나온다!

# 입력도 반환도 없는! 함수.... (아니 왜 만들어요,,,>?)
print('4. function example. no input... no return')
def say():
    print('hello world')
a=say()
print(a)
print('')
# 뭐... 어디든 쓰겠죠... 요래 만들고 요래 씁니다.

# 매개변수를 지정하자! 자! 아래와 같은 함수를 정의한다고 생각해봐요!
print('5. parameter designation. ')
def say(x,y):
    
    return [x+' hello world',y+' good bye']
    
# 이때 값을 아래와 같이 순차적으로 넣으면, 순차적으로 x와 y에 인수가 입력되겠죠?
results=say('ryu','kim')
print(results)

# 그런데 이 파라미터를 지정하여 인수를 입력할 수 있습니다!
results2=say(y='kim',x='ryu')
print(results2)
print('')
# 어때요? 직접 지정해서도 입력이 문제없이 진행되죠?
```

```python
# 입력값이 몇 개가 될지 모를 때 파라미터를 설정하는 법!
# 입력값이 단일인지 복수인지 모를때! 함수는 다음과 같이 정의한다!
print('6. parameter setting(don\'t know how many input).')
def add_many(*param):
    element=0
    for x in param:
        element=element+x
    return element
print('  test1. only one input: '+str(add_many(10)))
print('  test2. multi input: '+str(add_many(10,11,12,88,100,55)))
print('')

# 복합적으로 함수의 파라미터를 정의할수도 있다.
def operate(choice,*nums):
    element=0
    if choice=='plus':
        for x in nums:
            element=element+x
    elif choice=='minus':
        for x in nums:
            element=element-x
    else: element=0
            
    return element
print('  test3. plus input: '+str(operate('plus',10,11,12,88,100,55)))
print('  test4. minus input: '+str(operate('minus',11,12,88,100,55)))
print('')


# 키워드 매개변수: 함수를 정의할 때 매개변수를 키워드 매개변수로 설정할 수 있다. 그러면 입력값을 딕셔너리로 정리해준다.
# 예제를 보자!
print('7. keyword parameter setting')
def making_dic(**info):
    print(info)

making_dic(name='ryu',age=30,city='Daejeon')



# 키워드 매개변수와 가변 매개변수를 같이 쓰는 함수를 만들어 보면 다음과 같다.
print('8. mixed_function')
def mixed_info(name,*scores,**info): # 순서는 무조건 일반 매개변수, 가변 매개변수, 키워드 매개변수 순으로 진행할 것!!!
    print('='*25)
    print('='*5+'수상자 정보'+'='*5)
    print('='*25)
    print(f'이름: {name}')
    for key,value in info.items():
        print(f'{key}: {value}')
    print('='*25)
    total_score=0
    for x in scores:
        total_score=total_score+x
    print(f'total score: {total_score}')
    print('')
    return total_score

a_man=mixed_info('Ryu',50,60,100,height=180,weight=90,like='Bae')
b_man=mixed_info('Moon',50,60,100,20,80,70,height=200,weight=100,like='Bae',dislike='girl')
print('  test1. '+str(a_man))
print('  test2. '+str(b_man))
```

```python
# Return과 관련하여...
# Return은 하나이지만 복수의 return이 발생하면 아래와 같이 튜플로 값을 반환한다.
print('return test!')
def test_func(x,y):
    a=x+y
    b=x-y
    return a,b

test_1=test_func(50,60)
print('  test 1: '+str(test_1)) #튜플로 값 반환

# list의 형태로도 반환할 수 있다.
def test_func(x,y):
    a=x+y
    b=x-y
    return [a,b] # return 자체에 자료형을 부여하면 된다!

test_2=test_func(50,60)
print('  test 2: '+str(test_2)) #튜플로 값 반환

# 아예 하나씩 가져오는 방법도 있다. 함수의 출력을 바꿔보자
test_3,test_4=test_func(100,120)
print('  test 3: first value is '+str(test_3)+', second value is '+str(test_4)) #튜플로 값 반환

# 위와 같은 형태로 출력을 다양하게 취할 수 있다.

# 한편 return을 쓰면 함수를 바로 빠져나오기 떄문에 조건문에서 break?와 비슷하게 활용되어진다.
def return_test(choice):
    if choice=='plus':
        print('plus yesss')
    elif choice=='minus':
        return
        print('plus yesss')
    else: print('nothing')

print('test 1.')
return_test('')
print('test 2.')
return_test('plus')
print('test 3.')
return_test('minus') # 함수 안의 print가 실행되기 전에 함수를 빠져나온 것을 확인할 수 있다.
```

```python
# 함수 안에서 선언된 변수의 효력 범위
# 함수 안에서 선언된 변수는 함수 안에서만 유효하다.. 바깥에서는 유효하지 못하다...

def simple_test():
    good_chance=1
simple_test()
# print(good_chance) 이렇게 수행하면 정의되지 않은 변수라고 나온다.. 즉 정의된 함수 안에서 정의한 변수는 함수 안에서만 유효!

    
# 함수 안에서 함수 밖의 변수를 변경하는 방법이 있을까?
# 자! 함수 안에서 정의된 변수는 함수 안에서만 유효하다! 그런데! 함수로 함수 바깥의 변수를 바꾸고도 싶을 것이다.
# 이런 경우는 어떻게 하면 좋을까?
# 1. return 사용하기.. 이건 뭐 함수의 반환값으로 변수를 수정하는 것이니 당연! 예제는 생략한다.
# 2. global 사용하기
# global 변수는 함수에만 국한되는 local 변수가 아닌 코드 전체에 영향을 주는 변수이다. 즉 함수에서 변수를 global 변수로 지정하면 이는 함수를 빠져나가도 유효하다!

def simple_test():
    global good_chance
    good_chance=1

simple_test()
print(good_chance)

# 3. 리스트와 딕셔너리의 경우는 수정이 가능은 하다.. 우선 변경 가능한 자료형이기 때문이다.
def simple_test(listtt):
    listtt.append(66)
a=['Ryu',99]
print('before modify: '+str(a))
simple_test(a)
print('after modify: '+str(a))
# 이 점을 잘 활용해보는 것도 좋을듯.. 
```
