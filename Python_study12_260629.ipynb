```python
# 파일을 읽는 방법!!
# 1. readline
print('1. readline example')
f=open("/home/tester/work/python_study/file_ex_folder/test.txt",'r')
line=f.readline()
print(line)
f.close()
# 위의 코드를 실행하면 첫번째 줄의 데이터를 불러온다!
# 모든 줄의 데이터를 읽어오려면? while 문을 써서 반복하면 된다.
f=open("/home/tester/work/python_study/file_ex_folder/test.txt",'r')
while True:
    line=f.readline()
    if not line: break
    print(line)
f.close()
print('')

# 2. readlines
# 이 코드는 코드의 전체 내용을 읽어오며, 줄을 각 요소로 리스트를 반환하는 코드이다.
print('2. readlines example')
f=open("/home/tester/work/python_study/file_ex_folder/test.txt",'r')
lines=f.readlines()
print(lines)
f.close()
print('')

# 결과를 보면 끝 부분에 /n이 포함되어 있다.(줄바꿈) 이를 제거해야 되는 경우가 생기는데 strip 함수를 사용하면 쉽게 제거할 수 있다.
# 다만 전체로 할 수는 없고 줄당 한번씩 진행해줘야 해서 for 문을 활용하면 좋겠다.
f=open("/home/tester/work/python_study/file_ex_folder/test.txt",'r')
lines=f.readlines()
lines_new=[]
for x in lines:
    line=x.strip()
    print(line)
    lines_new.append(line)
f.close()
print('\n'+str(lines_new))
print('')

# 3. read 함수 사용하기.
# read를 사용하면 위와는 다르게 전체 내용을 그냥 하나의 문자열로 반환한다,
print('3. read example')
f=open("/home/tester/work/python_study/file_ex_folder/test.txt",'r')
data=f.read()
print(data)
f.close()

# 4. 파일 객체를 for문과 함께 사용하기
# f 객체를 직접 for에 활용하여 불러오는 것도 가능하다.
print('4. for문 활용하기')
line_new=[]
f=open("/home/tester/work/python_study/file_ex_folder/test.txt",'r')
for line in f:
    line_new.append(line.strip())
    print(line)
f.close()
print(str(line_new))
```

```python
# 추가하기 모드..
# r의 경우에는 값이 전체적으로 다시 쓰여진다.
# a를 활용하는 경우, 기존 파일에 값을 추가하게 된다.
print('add new data example')
f=open("/home/tester/work/python_study/file_ex_folder/test.txt",'a')
for i in range(10):
    data=f"{i+10}번째 값에 5를 곱하면? {(i+10)*5}! \n"
    f.write(data)
f.close()

print('read file after adding new line')
f=open("/home/tester/work/python_study/file_ex_folder/test.txt",'r')
data=f.read()
print(data)
f.close()
```

```python
# 중요사항!
# 파일은 열면 반드시 닫아주어야 한다. 위에서도 f.open 이후에 끝에는 반드시 f.close를 했다.

# with문에 대해서 알아본다.
with open("/home/tester/work/python_study/file_ex_folder/test.txt",'r') as f:
    new=f.read()
    print(new)

# with 문을 활용하면 파일을 따로 닫을 필요 없이 with 문을 빠져나오는 순간 객체 f가 자동으로 닫힌다.

# 객체가 닫혔는지 확인하는 방법은 file 객체의 내장함수 closed를 활용하면 된다.
print(f.closed)
# True면 닫혀있고, False면 열려있는 것..

# 참고로 with문에서 만들어진 변수는 해당 문을 벗어나도 변수로 활용할 수 있다. def 같은 정의된 함수와는 다름!
```
